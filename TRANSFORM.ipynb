```
# ============================================================
# PASO 4 - INVENTARIAR EL CONTENIDO DEL ZIP SIN DESCOMPRIMIR
# ============================================================

import zipfile
import os
from collections import Counter

ruta_zip = "/content/OneDrive_2026-06-05.zip"

print("=" * 90)
print("INVENTARIO DEL CONTENIDO DEL ZIP")
print("=" * 90)

with zipfile.ZipFile(ruta_zip, "r") as zip_ref:

    contenido = zip_ref.infolist()

    archivos = [
        item for item in contenido
        if not item.is_dir()
    ]

    carpetas = [
        item for item in contenido
        if item.is_dir()
    ]

    print(f"Cantidad de carpetas: {len(carpetas)}")
    print(f"Cantidad de archivos: {len(archivos)}")

    extensiones = Counter()

    for item in archivos:
        extension = os.path.splitext(item.filename)[1].lower()

        if extension == "":
            extension = "[sin extensión]"

        extensiones[extension] += 1

    print("\nARCHIVOS POR EXTENSIÓN")

    for extension, cantidad in extensiones.most_common():
        print(f"{extension}: {cantidad}")

    print("\nLISTADO DE ARCHIVOS")

    for numero, item in enumerate(archivos, start=1):
        tamaño_mb = item.file_size / (1024 * 1024)

        print(
            f"{numero:03d}. "
            f"{item.filename} | "
            f"{tamaño_mb:.2f} MB"
        )
```

```
# ============================================================
# PASO 5 - EXTRAER LOS ARCHIVOS ORIGINALES DEL ZIP
# ============================================================

import os
import zipfile
import shutil

# ------------------------------------------------------------
# RUTAS
# ------------------------------------------------------------

ruta_zip = "/content/OneDrive_2026-06-05.zip"
carpeta_extraccion = "/content/01_ORIGINALES"

print("=" * 90)
print("EXTRACCIÓN CONTROLADA DE LOS ARCHIVOS")
print("=" * 90)

# ------------------------------------------------------------
# LIMPIAR LA CARPETA SI YA EXISTÍA DE UNA EJECUCIÓN ANTERIOR
# ------------------------------------------------------------

if os.path.exists(carpeta_extraccion):
    shutil.rmtree(carpeta_extraccion)

os.makedirs(carpeta_extraccion, exist_ok=True)

# ------------------------------------------------------------
# EXTRAER EL ZIP
# ------------------------------------------------------------

with zipfile.ZipFile(ruta_zip, "r") as zip_ref:
    zip_ref.extractall(carpeta_extraccion)

print(f"Carpeta de extracción: {carpeta_extraccion}")

# ------------------------------------------------------------
# VERIFICAR LOS ARCHIVOS EXTRAÍDOS
# ------------------------------------------------------------

archivos_extraidos = []

for raiz, carpetas, archivos in os.walk(carpeta_extraccion):
    for archivo in archivos:
        ruta_completa = os.path.join(raiz, archivo)
        archivos_extraidos.append(ruta_completa)

print(f"Cantidad de archivos extraídos: {len(archivos_extraidos)}")

print("\nARCHIVOS EXTRAÍDOS")

for numero, ruta_completa in enumerate(
    sorted(archivos_extraidos),
    start=1
):
    ruta_relativa = os.path.relpath(
        ruta_completa,
        carpeta_extraccion
    )

    tamaño_mb = os.path.getsize(ruta_completa) / (1024 * 1024)

    print(
        f"{numero:03d}. "
        f"{ruta_relativa} | "
        f"{tamaño_mb:.2f} MB"
    )

# ------------------------------------------------------------
# VALIDACIÓN FINAL
# ------------------------------------------------------------

print("\n" + "=" * 90)

if len(archivos_extraidos) == 22:
    print("ESTADO: extracción correcta. Se encontraron los 22 archivos esperados.")
else:
    print(
        "ADVERTENCIA: la cantidad extraída no coincide con "
        "los 22 archivos esperados."
    )
```

```
# ============================================================
# PASO 6 - CLASIFICAR LOS ARCHIVOS POR GRUPO
# ============================================================

import os
import re
import pandas as pd

# ------------------------------------------------------------
# RUTA DE LOS ARCHIVOS EXTRAÍDOS
# ------------------------------------------------------------

carpeta_originales = "/content/01_ORIGINALES"

# ------------------------------------------------------------
# LISTAR TODOS LOS ARCHIVOS
# ------------------------------------------------------------

archivos = []

for raiz, carpetas, nombres_archivos in os.walk(carpeta_originales):
    for nombre_archivo in nombres_archivos:
        ruta_completa = os.path.join(raiz, nombre_archivo)

        archivos.append({
            "Archivo": nombre_archivo,
            "RutaCompleta": ruta_completa,
            "Extension": os.path.splitext(nombre_archivo)[1].lower()
        })

df_inventario = pd.DataFrame(archivos)

# ------------------------------------------------------------
# FUNCIÓN PARA CLASIFICAR CADA ARCHIVO
# ------------------------------------------------------------

def clasificar_archivo(nombre_archivo):

    nombre = nombre_archivo.lower()

    if "relacion cdp" in nombre or "relación cdp" in nombre:
        return "RELACION_CDP"

    elif "procesos por estado" in nombre:
        return "PROCESOS_ESTADO"

    elif re.search(r"\bpaa\s*20\d{2}", nombre):
        return "PAA"

    elif "ajustes y reservas" in nombre:
        return "AJUSTES_RESERVAS"

    elif "reporte de obras" in nombre:
        return "REPORTE_OBRAS"

    elif "reporte de pagos" in nombre:
        return "REPORTE_PAGOS"

    else:
        return "SIN_CLASIFICAR"

# ------------------------------------------------------------
# APLICAR CLASIFICACIÓN
# ------------------------------------------------------------

df_inventario["Grupo"] = df_inventario["Archivo"].apply(
    clasificar_archivo
)

# ------------------------------------------------------------
# EXTRAER AÑO DEL NOMBRE DEL ARCHIVO
# ------------------------------------------------------------

df_inventario["Anio"] = (
    df_inventario["Archivo"]
    .str.extract(r"(20\d{2})", expand=False)
)

# ------------------------------------------------------------
# ORDENAR RESULTADO
# ------------------------------------------------------------

df_inventario = df_inventario.sort_values(
    by=["Grupo", "Anio", "Archivo"]
).reset_index(drop=True)

# ------------------------------------------------------------
# MOSTRAR INVENTARIO
# ------------------------------------------------------------

print("=" * 100)
print("CLASIFICACIÓN DE ARCHIVOS")
print("=" * 100)

print(
    df_inventario[
        ["Grupo", "Anio", "Extension", "Archivo"]
    ].to_string(index=False)
)

# ------------------------------------------------------------
# RESUMEN POR GRUPO
# ------------------------------------------------------------

print("\n" + "=" * 100)
print("RESUMEN POR GRUPO")
print("=" * 100)

resumen = (
    df_inventario
    .groupby("Grupo")
    .size()
    .reset_index(name="CantidadArchivos")
)

print(resumen.to_string(index=False))

# ------------------------------------------------------------
# VALIDACIÓN FINAL
# ------------------------------------------------------------

sin_clasificar = df_inventario[
    df_inventario["Grupo"] == "SIN_CLASIFICAR"
]

print("\n" + "=" * 100)

if len(df_inventario) != 22:
    print(
        f"ADVERTENCIA: se encontraron {len(df_inventario)} archivos, "
        "pero se esperaban 22."
    )

elif not sin_clasificar.empty:
    print("ADVERTENCIA: existen archivos sin clasificar:")
    print(sin_clasificar["Archivo"].to_string(index=False))

else:
    print(
        "ESTADO: los 22 archivos fueron clasificados correctamente "
        "en sus grupos."
    )
```

```
# ============================================================
# PASO 7 - INSPECCIONAR LAS HOJAS DEL GRUPO RELACION_CDP
# ============================================================

import os
from openpyxl import load_workbook

# ------------------------------------------------------------
# SELECCIONAR ÚNICAMENTE LOS ARCHIVOS DEL GRUPO RELACION_CDP
# ------------------------------------------------------------

df_cdp = (
    df_inventario[
        df_inventario["Grupo"] == "RELACION_CDP"
    ]
    .sort_values("Anio")
    .reset_index(drop=True)
)

print("=" * 100)
print("INSPECCIÓN DE HOJAS - GRUPO RELACION_CDP")
print("=" * 100)

resultados_hojas = []
errores_hojas = []

# ------------------------------------------------------------
# ABRIR CADA ARCHIVO SIN CARGAR TODAS LAS CELDAS EN MEMORIA
# ------------------------------------------------------------

for _, fila in df_cdp.iterrows():

    archivo = fila["Archivo"]
    anio = fila["Anio"]
    ruta = fila["RutaCompleta"]

    print(f"\nArchivo: {archivo}")
    print(f"Año: {anio}")

    try:
        libro = load_workbook(
            filename=ruta,
            read_only=True,
            data_only=True
        )

        hojas = libro.worksheets

        print(f"Cantidad de hojas: {len(hojas)}")

        for numero, hoja in enumerate(hojas, start=1):

            estado = hoja.sheet_state

            print(
                f"  {numero}. "
                f"Hoja: {hoja.title} | "
                f"Estado: {estado}"
            )

            resultados_hojas.append({
                "Grupo": "RELACION_CDP",
                "Anio": anio,
                "Archivo": archivo,
                "NumeroHoja": numero,
                "NombreHoja": hoja.title,
                "EstadoHoja": estado
            })

        libro.close()

    except Exception as error:

        print(f"ERROR AL ABRIR EL ARCHIVO: {error}")

        errores_hojas.append({
            "Anio": anio,
            "Archivo": archivo,
            "Error": str(error)
        })

# ------------------------------------------------------------
# CREAR TABLA DE RESULTADOS
# ------------------------------------------------------------

df_hojas_cdp = pd.DataFrame(resultados_hojas)

print("\n" + "=" * 100)
print("RESUMEN DE HOJAS")
print("=" * 100)

if not df_hojas_cdp.empty:

    print(
        df_hojas_cdp[
            [
                "Anio",
                "NumeroHoja",
                "NombreHoja",
                "EstadoHoja",
                "Archivo"
            ]
        ].to_string(index=False)
    )

# ------------------------------------------------------------
# VALIDACIÓN FINAL
# ------------------------------------------------------------

print("\n" + "=" * 100)

if errores_hojas:

    print("ESTADO: algunos archivos presentaron errores.")

    for error in errores_hojas:
        print(
            f"- {error['Archivo']}: "
            f"{error['Error']}"
        )

elif len(df_cdp) != 4:

    print(
        f"ADVERTENCIA: se esperaban 4 archivos RELACION_CDP, "
        f"pero se encontraron {len(df_cdp)}."
    )

else:

    print(
        "ESTADO: los 4 archivos RELACION_CDP se abrieron "
        "correctamente y sus hojas fueron identificadas."
    )
```

```
# ============================================================
# PASO 9 - COMPARAR ENCABEZADOS DEL GRUPO RELACION_CDP
# ============================================================

import pandas as pd
import re
import unicodedata

# ------------------------------------------------------------
# FUNCIÓN DE NORMALIZACIÓN SOLO PARA COMPARAR
# NO MODIFICA TODAVÍA LOS NOMBRES ORIGINALES
# ------------------------------------------------------------

def normalizar_nombre_columna(nombre):

    nombre = str(nombre).strip()

    # Reemplazar saltos de línea por espacios
    nombre = nombre.replace("\n", " ")

    # Quitar tildes para facilitar la comparación
    nombre = unicodedata.normalize("NFKD", nombre)
    nombre = "".join(
        caracter
        for caracter in nombre
        if not unicodedata.combining(caracter)
    )

    # Convertir a minúsculas
    nombre = nombre.lower()

    # Reemplazar caracteres especiales por espacio
    nombre = re.sub(r"[^a-z0-9]+", " ", nombre)

    # Eliminar espacios repetidos
    nombre = re.sub(r"\s+", " ", nombre).strip()

    return nombre


print("=" * 120)
print("COMPARACIÓN DE COLUMNAS - RELACIÓN CDP")
print("=" * 120)

inventario_columnas = []
columnas_por_anio = {}
errores_columnas = []

# ------------------------------------------------------------
# LEER SOLO LOS ENCABEZADOS DE CADA ARCHIVO
# ------------------------------------------------------------

for _, fila in df_cdp.iterrows():

    anio = str(fila["Anio"])
    archivo = fila["Archivo"]
    ruta = fila["RutaCompleta"]
    hoja = hojas_principales_cdp[anio]

    print("\n" + "=" * 120)
    print(f"AÑO: {anio}")
    print(f"ARCHIVO: {archivo}")
    print(f"HOJA: {hoja}")
    print("=" * 120)

    try:

        encabezados = pd.read_excel(
            ruta,
            sheet_name=hoja,
            header=0,
            nrows=0,
            engine="openpyxl"
        )

        columnas_originales = list(encabezados.columns)

        columnas_por_anio[anio] = {
            normalizar_nombre_columna(columna)
            for columna in columnas_originales
        }

        print(f"Cantidad de columnas: {len(columnas_originales)}\n")

        for posicion, columna in enumerate(
            columnas_originales,
            start=1
        ):

            columna_normalizada = normalizar_nombre_columna(
                columna
            )

            print(
                f"{posicion:02d}. "
                f"Original: {columna} | "
                f"Normalizada: {columna_normalizada}"
            )

            inventario_columnas.append({
                "Anio": anio,
                "Archivo": archivo,
                "Posicion": posicion,
                "ColumnaOriginal": str(columna),
                "ColumnaNormalizada": columna_normalizada
            })

    except Exception as error:

        print(f"ERROR: {error}")

        errores_columnas.append({
            "Anio": anio,
            "Archivo": archivo,
            "Error": str(error)
        })

# ------------------------------------------------------------
# CREAR DATAFRAME DE INVENTARIO
# ------------------------------------------------------------

df_columnas_cdp = pd.DataFrame(inventario_columnas)

# ------------------------------------------------------------
# IDENTIFICAR COLUMNAS COMUNES A LOS CUATRO AÑOS
# ------------------------------------------------------------

print("\n" + "=" * 120)
print("COLUMNAS NORMALIZADAS COMUNES A LOS CUATRO AÑOS")
print("=" * 120)

if len(columnas_por_anio) == 4:

    columnas_comunes = set.intersection(
        *columnas_por_anio.values()
    )

    if columnas_comunes:

        for columna in sorted(columnas_comunes):
            print(f"- {columna}")

    else:
        print(
            "No existen columnas con exactamente el mismo "
            "nombre normalizado en los cuatro años."
        )

# ------------------------------------------------------------
# MOSTRAR COLUMNAS EXCLUSIVAS O DIFERENTES POR AÑO
# ------------------------------------------------------------

print("\n" + "=" * 120)
print("COLUMNAS NO COMUNES POR AÑO")
print("=" * 120)

if len(columnas_por_anio) == 4:

    for anio in sorted(columnas_por_anio):

        otros_anios = [
            columnas
            for otro_anio, columnas in columnas_por_anio.items()
            if otro_anio != anio
        ]

        union_otros = set.union(*otros_anios)

        exclusivas = (
            columnas_por_anio[anio]
            - union_otros
        )

        print(f"\nAño {anio}:")

        if exclusivas:

            for columna in sorted(exclusivas):
                print(f"- {columna}")

        else:
            print("- No tiene columnas exclusivas.")

# ------------------------------------------------------------
# MATRIZ DE PRESENCIA POR AÑO
# ------------------------------------------------------------

print("\n" + "=" * 120)
print("MATRIZ DE PRESENCIA DE COLUMNAS NORMALIZADAS")
print("=" * 120)

if columnas_por_anio:

    todas_las_columnas = sorted(
        set.union(*columnas_por_anio.values())
    )

    matriz_presencia = pd.DataFrame({
        "ColumnaNormalizada": todas_las_columnas
    })

    for anio in sorted(columnas_por_anio):

        matriz_presencia[anio] = (
            matriz_presencia["ColumnaNormalizada"]
            .isin(columnas_por_anio[anio])
            .map({
                True: "Sí",
                False: "No"
            })
        )

    print(
        matriz_presencia.to_string(
            index=False
        )
    )

# ------------------------------------------------------------
# VALIDACIÓN FINAL
# ------------------------------------------------------------

print("\n" + "=" * 120)

if errores_columnas:

    print(
        "ESTADO: se presentaron errores al leer "
        "los encabezados."
    )

    for error in errores_columnas:
        print(
            f"- Año {error['Anio']} | "
            f"{error['Archivo']} | "
            f"{error['Error']}"
        )

elif len(columnas_por_anio) != 4:

    print(
        "ADVERTENCIA: no se obtuvieron encabezados "
        "de los cuatro años."
    )

else:

    print(
        "ESTADO: los encabezados de los cuatro años "
        "fueron comparados correctamente."
    )
```

```
# ============================================================
# PASO 12 - IDENTIFICAR LAS FILAS REALES DE NEGOCIO
# GRUPO RELACION_CDP
# ============================================================

import pandas as pd
import numpy as np
import re
import unicodedata


# ------------------------------------------------------------
# FUNCIÓN PARA NORMALIZAR ENCABEZADOS SOLO PARA BÚSQUEDA
# ------------------------------------------------------------

def normalizar_encabezado(nombre):

    nombre = str(nombre).strip()
    nombre = nombre.replace("\n", " ")

    nombre = unicodedata.normalize("NFKD", nombre)

    nombre = "".join(
        caracter
        for caracter in nombre
        if not unicodedata.combining(caracter)
    )

    nombre = nombre.lower()
    nombre = re.sub(r"[^a-z0-9]+", " ", nombre)
    nombre = re.sub(r"\s+", " ", nombre).strip()

    return nombre


# ------------------------------------------------------------
# CAMPOS CLAVE POR AÑO
# ------------------------------------------------------------

campos_clave_cdp = {

    "2023": {
        "Fecha": "FECHA",
        "Proyecto": "PROYECTO PEP",
        "NumeroCDP": "No CDP"
    },

    "2024": {
        "Fecha": "FECHA CDP",
        "Proyecto": "PROYECTO PEP",
        "NumeroCDP": "No CDP"
    },

    "2025": {
        "Fecha": "Fecha solicitud CDP",
        "Proyecto": "Proyecto",
        "NumeroCDP": "N° CDP"
    },

    "2026": {
        "Fecha": "Fecha solicitud CDP",
        "Proyecto": "Proyecto",
        "NumeroCDP": "N° CDP"
    }
}


# ------------------------------------------------------------
# FUNCIÓN PARA LOCALIZAR UNA COLUMNA SIN DEPENDER DE ESPACIOS
# ------------------------------------------------------------

def encontrar_columna(df, nombre_buscado):

    nombre_normalizado = normalizar_encabezado(nombre_buscado)

    coincidencias = [
        columna
        for columna in df.columns
        if normalizar_encabezado(columna) == nombre_normalizado
    ]

    if not coincidencias:
        return None

    return coincidencias[0]


# ------------------------------------------------------------
# CONTENEDORES
# ------------------------------------------------------------

resumen_filas_reales = []
filas_parciales_detectadas = []
datos_reales_por_anio = {}
errores_filas_reales = []


print("=" * 130)
print("IDENTIFICACIÓN DE FILAS REALES DE NEGOCIO - RELACIÓN CDP")
print("=" * 130)


# ------------------------------------------------------------
# ANALIZAR CADA AÑO
# ------------------------------------------------------------

for _, fila_inventario in df_cdp.iterrows():

    anio = str(fila_inventario["Anio"])
    archivo = fila_inventario["Archivo"]

    print("\n" + "=" * 130)
    print(f"AÑO: {anio}")
    print(f"ARCHIVO: {archivo}")
    print("=" * 130)

    try:

        # Utilizamos el DataFrame ya leído en el Paso 11
        df_anio = datos_por_anio[anio].copy()

        # Mantener el número real de fila del Excel
        # Encabezado = fila 1, primer dato = fila 2
        df_anio["_FilaExcel"] = df_anio.index + 2

        columna_fecha = encontrar_columna(
            df_anio,
            campos_clave_cdp[anio]["Fecha"]
        )

        columna_proyecto = encontrar_columna(
            df_anio,
            campos_clave_cdp[anio]["Proyecto"]
        )

        columna_cdp = encontrar_columna(
            df_anio,
            campos_clave_cdp[anio]["NumeroCDP"]
        )

        if not all([
            columna_fecha,
            columna_proyecto,
            columna_cdp
        ]):
            raise ValueError(
                "No fue posible localizar todos los campos clave."
            )

        # ----------------------------------------------------
        # LIMPIAR CADENAS VACÍAS EN LOS CAMPOS CLAVE
        # ----------------------------------------------------

        for columna in [
            columna_fecha,
            columna_proyecto,
            columna_cdp
        ]:

            df_anio[columna] = df_anio[columna].replace(
                r"^\s*$",
                np.nan,
                regex=True
            )

        # ----------------------------------------------------
        # INDICADORES DE PRESENCIA
        # ----------------------------------------------------

        tiene_fecha = df_anio[columna_fecha].notna()
        tiene_proyecto = df_anio[columna_proyecto].notna()
        tiene_cdp = df_anio[columna_cdp].notna()

        cantidad_claves = (
            tiene_fecha.astype(int)
            + tiene_proyecto.astype(int)
            + tiene_cdp.astype(int)
        )

        # Registro candidato:
        # tiene al menos uno de los tres campos clave
        mascara_registro_candidato = cantidad_claves >= 1

        df_candidatos = (
            df_anio.loc[mascara_registro_candidato]
            .copy()
        )

        # Registro completo:
        # tiene los tres campos clave
        mascara_registro_completo = cantidad_claves == 3

        # Registro parcial:
        # tiene uno o dos campos clave
        mascara_registro_parcial = (
            (cantidad_claves >= 1)
            &
            (cantidad_claves < 3)
        )

        df_parciales = (
            df_anio.loc[
                mascara_registro_parcial,
                [
                    "_FilaExcel",
                    columna_fecha,
                    columna_proyecto,
                    columna_cdp
                ]
            ]
            .copy()
        )

        # Guardar candidatos para pasos posteriores,
        # pero todavía no se consolidan
        datos_reales_por_anio[anio] = df_candidatos.copy()

        cantidad_original = len(df_anio)
        cantidad_candidatos = len(df_candidatos)
        cantidad_completos = int(
            mascara_registro_completo.sum()
        )
        cantidad_parciales = int(
            mascara_registro_parcial.sum()
        )
        cantidad_residuales = (
            cantidad_original - cantidad_candidatos
        )

        primera_fila_real = (
            int(df_candidatos["_FilaExcel"].min())
            if not df_candidatos.empty
            else None
        )

        ultima_fila_real = (
            int(df_candidatos["_FilaExcel"].max())
            if not df_candidatos.empty
            else None
        )

        print(f"Filas leídas originalmente: {cantidad_original}")
        print(f"Registros candidatos: {cantidad_candidatos}")
        print(f"Registros con las 3 claves: {cantidad_completos}")
        print(f"Registros parciales: {cantidad_parciales}")
        print(f"Filas residuales descartables: {cantidad_residuales}")
        print(f"Primera fila real del Excel: {primera_fila_real}")
        print(f"Última fila real del Excel: {ultima_fila_real}")

        # ----------------------------------------------------
        # DISTRIBUCIÓN POR COMBINACIÓN DE CAMPOS CLAVE
        # ----------------------------------------------------

        combinaciones = pd.DataFrame({
            "TieneFecha": tiene_fecha,
            "TieneProyecto": tiene_proyecto,
            "TieneNumeroCDP": tiene_cdp
        })

        resumen_combinaciones = (
            combinaciones
            .value_counts()
            .reset_index(name="CantidadFilas")
            .sort_values(
                "CantidadFilas",
                ascending=False
            )
        )

        print("\nCOMBINACIONES DE CAMPOS CLAVE")

        print(
            resumen_combinaciones.to_string(
                index=False
            )
        )

        # ----------------------------------------------------
        # MOSTRAR FILAS PARCIALES
        # ----------------------------------------------------

        print("\nREGISTROS PARCIALES")

        if df_parciales.empty:

            print(
                "No se encontraron registros parciales. "
                "Todos los candidatos tienen las tres claves."
            )

        else:

            print(
                df_parciales.head(30).to_string(
                    index=False
                )
            )

            for _, fila_parcial in df_parciales.iterrows():

                filas_parciales_detectadas.append({
                    "Anio": anio,
                    "Archivo": archivo,
                    "FilaExcel": fila_parcial["_FilaExcel"],
                    "Fecha": fila_parcial[columna_fecha],
                    "Proyecto": fila_parcial[columna_proyecto],
                    "NumeroCDP": fila_parcial[columna_cdp]
                })

        # ----------------------------------------------------
        # MOSTRAR LAS ÚLTIMAS FILAS CANDIDATAS
        # ----------------------------------------------------

        print("\nÚLTIMAS 10 FILAS CANDIDATAS")

        columnas_muestra = [
            "_FilaExcel",
            columna_fecha,
            columna_proyecto,
            columna_cdp
        ]

        print(
            df_candidatos[
                columnas_muestra
            ]
            .tail(10)
            .to_string(index=False)
        )

        resumen_filas_reales.append({
            "Anio": anio,
            "Archivo": archivo,
            "FilasOriginales": cantidad_original,
            "RegistrosCandidatos": cantidad_candidatos,
            "RegistrosCompletos": cantidad_completos,
            "RegistrosParciales": cantidad_parciales,
            "FilasResiduales": cantidad_residuales,
            "PrimeraFilaExcel": primera_fila_real,
            "UltimaFilaExcel": ultima_fila_real
        })

    except Exception as error:

        print(f"ERROR: {error}")

        errores_filas_reales.append({
            "Anio": anio,
            "Archivo": archivo,
            "Error": str(error)
        })


# ------------------------------------------------------------
# RESUMEN GENERAL
# ------------------------------------------------------------

df_resumen_filas_reales = pd.DataFrame(
    resumen_filas_reales
)

df_filas_parciales_cdp = pd.DataFrame(
    filas_parciales_detectadas
)

print("\n" + "=" * 130)
print("RESUMEN GENERAL DE FILAS REALES")
print("=" * 130)

if not df_resumen_filas_reales.empty:

    print(
        df_resumen_filas_reales.to_string(
            index=False
        )
    )


# ------------------------------------------------------------
# VALIDACIÓN FINAL
# ------------------------------------------------------------

print("\n" + "=" * 130)

if errores_filas_reales:

    print(
        "ESTADO: se presentaron errores al identificar "
        "las filas reales."
    )

    for error in errores_filas_reales:

        print(
            f"- Año {error['Anio']} | "
            f"{error['Archivo']} | "
            f"{error['Error']}"
        )

elif len(datos_reales_por_anio) != 4:

    print(
        "ADVERTENCIA: no se analizaron correctamente "
        "los cuatro años."
    )

else:

    print(
        "ESTADO: se identificaron las filas candidatas "
        "de los cuatro años. Todavía no se eliminaron, "
        "renombraron ni unieron datos."
    )
```

```
# ============================================================
# PASO 15 - SELECCIONAR Y HOMOLOGAR COLUMNAS POR AÑO
# GRUPO RELACION_CDP
# ============================================================

import pandas as pd
import numpy as np
import re
import unicodedata


# ------------------------------------------------------------
# FUNCIÓN PARA NORMALIZAR ENCABEZADOS DURANTE LA BÚSQUEDA
# ------------------------------------------------------------

def normalizar_encabezado(nombre):

    nombre = str(nombre).strip()
    nombre = nombre.replace("\n", " ")

    nombre = unicodedata.normalize("NFKD", nombre)

    nombre = "".join(
        caracter
        for caracter in nombre
        if not unicodedata.combining(caracter)
    )

    nombre = nombre.lower()
    nombre = re.sub(r"[^a-z0-9]+", " ", nombre)
    nombre = re.sub(r"\s+", " ", nombre).strip()

    return nombre


# ------------------------------------------------------------
# FUNCIÓN PARA ENCONTRAR EL NOMBRE REAL DE UNA COLUMNA
# ------------------------------------------------------------

def encontrar_columna_real(df, nombre_buscado):

    if nombre_buscado is None:
        return None

    nombre_normalizado = normalizar_encabezado(
        nombre_buscado
    )

    coincidencias = [
        columna
        for columna in df.columns
        if normalizar_encabezado(columna)
        == nombre_normalizado
    ]

    if not coincidencias:
        return None

    return coincidencias[0]


# ------------------------------------------------------------
# COLUMNAS APROBADAS Y ORDEN FINAL
# ------------------------------------------------------------

columnas_aprobadas_cdp = (
    df_matriz_homologacion_cdp[
        df_matriz_homologacion_cdp["Usar"] == "SI"
    ]["NombreEstandar"]
    .tolist()
)


# ------------------------------------------------------------
# FUNCIÓN PARA HOMOLOGAR UN AÑO
# ------------------------------------------------------------

def homologar_cdp_por_anio(
    df_origen,
    anio,
    archivo_origen
):

    columna_matriz = f"Columna{anio}"

    resultado = pd.DataFrame(
        index=df_origen.index
    )

    columnas_no_encontradas = []

    for _, regla in (
        df_matriz_homologacion_cdp[
            df_matriz_homologacion_cdp["Usar"] == "SI"
        ].iterrows()
    ):

        nombre_estandar = regla["NombreEstandar"]
        nombre_original = regla[columna_matriz]

        # ----------------------------------------------------
        # SI EL CAMPO NO EXISTE EN ESE AÑO, CREARLO NULO
        # ----------------------------------------------------

        if pd.isna(nombre_original) or nombre_original is None:

            resultado[nombre_estandar] = pd.NA

            continue

        # ----------------------------------------------------
        # LOCALIZAR LA COLUMNA REAL
        # ----------------------------------------------------

        columna_real = encontrar_columna_real(
            df_origen,
            nombre_original
        )

        if columna_real is None:

            resultado[nombre_estandar] = pd.NA

            columnas_no_encontradas.append({
                "Anio": anio,
                "NombreEstandar": nombre_estandar,
                "NombreOriginalEsperado": nombre_original
            })

        else:

            resultado[nombre_estandar] = (
                df_origen[columna_real]
            )

    # --------------------------------------------------------
    # COLUMNAS DE TRAZABILIDAD
    # --------------------------------------------------------

    resultado["AnioFuente"] = int(anio)
    resultado["ArchivoFuente"] = archivo_origen

    if "_FilaExcel" in df_origen.columns:
        resultado["FilaExcelFuente"] = (
            df_origen["_FilaExcel"]
            .astype("Int64")
        )
    else:
        resultado["FilaExcelFuente"] = pd.NA

    return resultado, columnas_no_encontradas


# ------------------------------------------------------------
# PROCESAR LOS CUATRO AÑOS SIN UNIRLOS
# ------------------------------------------------------------

datos_homologados_por_anio = {}
columnas_faltantes_homologacion = []

print("=" * 150)
print("HOMOLOGACIÓN ESTRUCTURAL POR AÑO - RELACIÓN CDP")
print("=" * 150)


for _, fila_inventario in df_cdp.iterrows():

    anio = str(fila_inventario["Anio"])
    archivo = fila_inventario["Archivo"]

    print("\n" + "=" * 150)
    print(f"AÑO: {anio}")
    print(f"ARCHIVO: {archivo}")
    print("=" * 150)

    # Datos reales definidos en el Paso 12
    df_real = datos_reales_por_anio[anio].copy()

    df_homologado, faltantes = homologar_cdp_por_anio(
        df_origen=df_real,
        anio=anio,
        archivo_origen=archivo
    )

    datos_homologados_por_anio[anio] = (
        df_homologado.copy()
    )

    columnas_faltantes_homologacion.extend(
        faltantes
    )

    print(f"Registros de entrada: {len(df_real)}")
    print(f"Registros homologados: {len(df_homologado)}")
    print(
        f"Columnas de negocio homologadas: "
        f"{len(columnas_aprobadas_cdp)}"
    )
    print(
        f"Columnas de trazabilidad: 3"
    )
    print(
        f"Total de columnas del resultado: "
        f"{len(df_homologado.columns)}"
    )

    print("\nCOLUMNAS RESULTANTES")

    for numero, columna in enumerate(
        df_homologado.columns,
        start=1
    ):
        print(f"{numero:02d}. {columna}")

    print("\nPRIMEROS 3 REGISTROS")

    columnas_muestra = [
        "FechaSolicitudCDP",
        "Proyecto",
        "DescripcionProyecto",
        "NumeroCDP",
        "NumeroContrato",
        "NumeroRP",
        "ValorCDP",
        "AnioFuente",
        "FilaExcelFuente"
    ]

    print(
        df_homologado[
            columnas_muestra
        ]
        .head(3)
        .to_string(
            index=False,
            max_colwidth=50
        )
    )


# ------------------------------------------------------------
# CREAR VARIABLES INDIVIDUALES POR AÑO
# ------------------------------------------------------------

df_cdp_2023_homologado = (
    datos_homologados_por_anio["2023"].copy()
)

df_cdp_2024_homologado = (
    datos_homologados_por_anio["2024"].copy()
)

df_cdp_2025_homologado = (
    datos_homologados_por_anio["2025"].copy()
)

df_cdp_2026_homologado = (
    datos_homologados_por_anio["2026"].copy()
)


# ------------------------------------------------------------
# VALIDAR ESTRUCTURA IDÉNTICA ENTRE LOS CUATRO AÑOS
# ------------------------------------------------------------

print("\n" + "=" * 150)
print("VALIDACIÓN DE ESTRUCTURA ENTRE AÑOS")
print("=" * 150)

columnas_2023 = list(
    df_cdp_2023_homologado.columns
)

estructura_identica = all(
    list(df_anio.columns) == columnas_2023
    for df_anio in datos_homologados_por_anio.values()
)

for anio, df_anio in datos_homologados_por_anio.items():

    print(
        f"Año {anio}: "
        f"{len(df_anio)} registros | "
        f"{len(df_anio.columns)} columnas"
    )


# ------------------------------------------------------------
# VALIDAR CONTEO ESPERADO
# ------------------------------------------------------------

conteos_esperados = {
    "2023": 249,
    "2024": 313,
    "2025": 549,
    "2026": 231
}

conteos_correctos = all(
    len(datos_homologados_por_anio[anio])
    == cantidad_esperada
    for anio, cantidad_esperada
    in conteos_esperados.items()
)


# ------------------------------------------------------------
# MOSTRAR COLUMNAS QUE NO PUDIERON LOCALIZARSE
# ------------------------------------------------------------

print("\n" + "=" * 150)
print("COLUMNAS ESPERADAS QUE NO FUERON LOCALIZADAS")
print("=" * 150)

if columnas_faltantes_homologacion:

    df_columnas_faltantes = pd.DataFrame(
        columnas_faltantes_homologacion
    )

    print(
        df_columnas_faltantes.to_string(
            index=False
        )
    )

else:

    print(
        "Todas las columnas existentes por año "
        "fueron localizadas correctamente."
    )


# ------------------------------------------------------------
# VALIDACIÓN FINAL
# ------------------------------------------------------------

print("\n" + "=" * 150)

if columnas_faltantes_homologacion:

    print(
        "ADVERTENCIA: existen columnas esperadas "
        "que no fueron localizadas."
    )

elif not estructura_identica:

    print(
        "ADVERTENCIA: los cuatro DataFrames no tienen "
        "exactamente la misma estructura."
    )

elif not conteos_correctos:

    print(
        "ADVERTENCIA: alguno de los años no conserva "
        "la cantidad esperada de registros."
    )

else:

    print(
        "ESTADO: los cuatro años fueron homologados "
        "correctamente, conservando sus registros y con "
        "la misma estructura. Todavía no fueron unidos "
        "ni se aplicaron limpiezas de contenido."
    )
```

```
# ============================================================
# PASO 17 CORREGIDO - PROBAR FUNCIONES DE LIMPIEZA
# SOBRE VALORES DE MUESTRA
# ============================================================

import pandas as pd
import re
import math


# ------------------------------------------------------------
# IDENTIFICAR VALORES VACÍOS
# ------------------------------------------------------------

def es_valor_vacio(valor):

    if valor is None:
        return True

    try:
        if pd.isna(valor):
            return True
    except (TypeError, ValueError):
        pass

    return str(valor).strip() == ""


# ------------------------------------------------------------
# LIMPIAR CÓDIGOS
# ------------------------------------------------------------

def limpiar_codigo(
    valor,
    conservar_varios=True,
    minimo_digitos=5
):
    """
    Limpia campos que deben contener códigos numéricos.

    Reglas:
    1. Vacío -> pd.NA.
    2. Elimina la terminación .0.
    3. Extrae códigos de mínimo 5 dígitos.
    4. No captura años de 4 dígitos ni números cortos.
    5. Puede conservar varios códigos separados por |.
    6. Elimina códigos repetidos conservando el orden.
    """

    if es_valor_vacio(valor):
        return pd.NA

    # Números leídos directamente por pandas
    if isinstance(valor, int):
        texto = str(valor)

    elif isinstance(valor, float):

        if math.isnan(valor):
            return pd.NA

        if valor.is_integer():
            texto = str(int(valor))
        else:
            texto = str(valor)

    else:
        texto = str(valor).strip()

    # Quitar .0 de números enteros representados como texto
    texto = re.sub(
        r"(?<=\d)\.0(?=\s|$)",
        "",
        texto
    )

    # Extraer únicamente secuencias con el mínimo de dígitos
    patron = rf"(?<!\d)\d{{{minimo_digitos},}}(?!\d)"

    codigos = re.findall(
        patron,
        texto
    )

    if not codigos:
        return pd.NA

    # Quitar duplicados manteniendo el orden
    codigos_unicos = list(
        dict.fromkeys(codigos)
    )

    if conservar_varios:
        return "|".join(codigos_unicos)

    return codigos_unicos[0]


# ------------------------------------------------------------
# LIMPIAR VALORES NUMÉRICOS
# ------------------------------------------------------------

def limpiar_numero(valor):
    """
    Limpia campos que deben contener un único valor numérico.

    Solo acepta valores completamente numéricos.
    Los textos descriptivos quedan como pd.NA.
    """

    if es_valor_vacio(valor):
        return pd.NA

    if isinstance(valor, (int, float)):

        try:
            if pd.isna(valor):
                return pd.NA
        except (TypeError, ValueError):
            pass

        if isinstance(valor, float) and valor.is_integer():
            return int(valor)

        return valor

    texto = str(valor).strip()

    if re.fullmatch(
        r"-?\d+(?:[.,]\d+)?",
        texto
    ):

        texto = texto.replace(",", ".")

        numero = float(texto)

        if numero.is_integer():
            return int(numero)

        return numero

    return pd.NA


# ------------------------------------------------------------
# CASOS DE PRUEBA DE CÓDIGOS
# ------------------------------------------------------------

casos_codigo = [
    "4000115729",
    "4000115729\nSolicitud anulación el 20 de diciembre de 2023",
    "4600106201\nCONVENIO AMVA No. 413",
    "5000273272\n5000273542\n5000274798",
    4600094136.0,
    "No aplica",
    "SALDO LIBERADO",
    "Convenio 413 del año 2023",
    None,
    "",
    "   "
]


# ------------------------------------------------------------
# CASOS DE PRUEBA NUMÉRICOS
# ------------------------------------------------------------

casos_numero = [
    3453802.0,
    "3453802",
    "3453802.50",
    "3453802,50",
    "SALDO LIBERADO",
    "10 días",
    "No aplica",
    None,
    ""
]


# ------------------------------------------------------------
# MOSTRAR PRUEBAS DE CÓDIGOS
# ------------------------------------------------------------

print("=" * 120)
print("PRUEBA CORREGIDA DE LIMPIEZA DE CÓDIGOS")
print("=" * 120)

resultados_codigos = []

for valor in casos_codigo:

    resultados_codigos.append({
        "ValorOriginal": repr(valor),

        "ConservarVarios": limpiar_codigo(
            valor,
            conservar_varios=True
        ),

        "SoloPrimerCodigo": limpiar_codigo(
            valor,
            conservar_varios=False
        )
    })

df_prueba_codigos = pd.DataFrame(
    resultados_codigos
)

print(
    df_prueba_codigos.to_string(
        index=False
    )
)


# ------------------------------------------------------------
# MOSTRAR PRUEBAS NUMÉRICAS
# ------------------------------------------------------------

print("\n" + "=" * 120)
print("PRUEBA DE LIMPIEZA DE VALORES NUMÉRICOS")
print("=" * 120)

resultados_numeros = []

for valor in casos_numero:

    resultado = limpiar_numero(valor)

    resultados_numeros.append({
        "ValorOriginal": repr(valor),
        "Resultado": resultado,
        "TipoResultado": (
            type(resultado).__name__
            if not pd.isna(resultado)
            else "NA"
        )
    })

df_prueba_numeros = pd.DataFrame(
    resultados_numeros
)

print(
    df_prueba_numeros.to_string(
        index=False
    )
)


# ------------------------------------------------------------
# VALIDACIONES AUTOMÁTICAS CORREGIDAS
# ------------------------------------------------------------

validaciones = {

    "Extrae CDP y no captura el año":
        limpiar_codigo(
            "4000115729\nSolicitud anulación el 20 de diciembre de 2023"
        ) == "4000115729",

    "Extrae contrato y no captura número corto de convenio":
        limpiar_codigo(
            "4600106201\nCONVENIO AMVA No. 413"
        ) == "4600106201",

    "Conserva varios RP":
        limpiar_codigo(
            "5000273272\n5000273542\n5000274798"
        ) == "5000273272|5000273542|5000274798",

    "Elimina decimal cero":
        limpiar_codigo(
            4600094136.0
        ) == "4600094136",

    "Texto sin código válido queda nulo":
        pd.isna(
            limpiar_codigo("No aplica")
        ),

    "Año y número corto no se toman como códigos":
        pd.isna(
            limpiar_codigo(
                "Convenio 413 del año 2023"
            )
        ),

    "Saldo liberado queda nulo":
        pd.isna(
            limpiar_numero("SALDO LIBERADO")
        ),

    "Plazo con texto queda nulo":
        pd.isna(
            limpiar_numero("10 días")
        )
}


print("\n" + "=" * 120)
print("VALIDACIONES AUTOMÁTICAS CORREGIDAS")
print("=" * 120)

todas_correctas = True

for validacion, resultado in validaciones.items():

    estado = "OK" if resultado else "ERROR"

    print(f"{estado}: {validacion}")

    if not resultado:
        todas_correctas = False


print("\n" + "=" * 120)

if todas_correctas:

    print(
        "ESTADO: todas las reglas corregidas fueron "
        "validadas correctamente. No se capturan años "
        "ni números cortos como códigos."
    )

else:

    print(
        "ESTADO: alguna regla continúa generando "
        "un resultado incorrecto. No se debe avanzar."
    )
```

```
# ============================================================
# PASO 18 - APLICAR LIMPIEZA CONTROLADA POR AÑO
# GRUPO RELACION_CDP
# ============================================================

import pandas as pd
import numpy as np
import re


# ------------------------------------------------------------
# CONFIGURACIÓN DE COLUMNAS
# ------------------------------------------------------------

# Campos de código:
# se conserva únicamente el primer código válido
columnas_codigo_unico = [
    "Proyecto",
    "Fondo",
    "PosicionPresupuestal",
    "NumeroEstudioPrevio",
    "SolicitudPedido",
    "NumeroContrato",
    "NumeroCDP"
]

# NumeroRP puede contener varios códigos legítimos
columnas_codigo_multiple = [
    "NumeroRP"
]

# Campos que deben quedar numéricos
columnas_numericas = [
    "ValorCDP",
    "SaldoCDP",
    "ValorContrato",
    "PlazoMeses",
    "PlazoDias"
]

# Campos descriptivos
columnas_texto = [
    "DescripcionProyecto",
    "DescripcionFondo",
    "DescripcionClasificador",
    "ActividadAnexoPresupuestalMGA",
    "ObjetoCDP",
    "Subsecretaria",
    "Responsable",
    "Contratista",
    "ObservacionAnulacion",
    "ArchivoFuente"
]


# ------------------------------------------------------------
# FUNCIÓN PARA LIMPIAR CAMPOS DE TEXTO
# ------------------------------------------------------------

def limpiar_texto(valor):
    """
    Limpia únicamente espacios exteriores y cadenas vacías.

    No cambia mayúsculas, minúsculas, tildes ni el contenido
    descriptivo del campo.
    """

    if es_valor_vacio(valor):
        return pd.NA

    texto = str(valor).strip()

    # Normalizar saltos de línea de Windows y Mac
    texto = texto.replace("\r\n", "\n")
    texto = texto.replace("\r", "\n")

    # Eliminar espacios antes y después de cada línea,
    # conservando los saltos internos
    lineas = [
        linea.strip()
        for linea in texto.split("\n")
    ]

    texto = "\n".join(lineas).strip()

    if texto == "":
        return pd.NA

    return texto


# ------------------------------------------------------------
# FUNCIÓN DE LIMPIEZA PARA UN DATAFRAME
# ------------------------------------------------------------

def limpiar_dataframe_cdp(df_origen, anio):
    """
    Aplica las reglas aprobadas a una copia del DataFrame.

    No altera el DataFrame original.
    """

    df_limpio = df_origen.copy()

    resumen_cambios = []
    muestras_cambios = []

    # --------------------------------------------------------
    # FECHA
    # --------------------------------------------------------

    columna = "FechaSolicitudCDP"

    valores_antes = df_limpio[columna].copy()

    df_limpio[columna] = pd.to_datetime(
        df_limpio[columna],
        errors="coerce"
    )

    nulos_antes = int(valores_antes.isna().sum())
    nulos_despues = int(df_limpio[columna].isna().sum())

    resumen_cambios.append({
        "Anio": anio,
        "Columna": columna,
        "TipoLimpieza": "Fecha",
        "ConDatoAntes": int(valores_antes.notna().sum()),
        "ConDatoDespues": int(df_limpio[columna].notna().sum()),
        "NulosNuevos": nulos_despues - nulos_antes
    })

    # --------------------------------------------------------
    # CÓDIGOS ÚNICOS
    # --------------------------------------------------------

    for columna in columnas_codigo_unico:

        valores_antes = df_limpio[columna].copy()

        df_limpio[columna] = (
            df_limpio[columna]
            .apply(
                lambda valor: limpiar_codigo(
                    valor,
                    conservar_varios=False,
                    minimo_digitos=5
                )
            )
            .astype("string")
        )

        mascara_cambio = (
            valores_antes.fillna("<NA>").astype(str)
            !=
            df_limpio[columna].fillna("<NA>").astype(str)
        )

        nulos_antes = int(
            valores_antes.apply(es_valor_vacio).sum()
        )

        nulos_despues = int(
            df_limpio[columna].isna().sum()
        )

        resumen_cambios.append({
            "Anio": anio,
            "Columna": columna,
            "TipoLimpieza": "CodigoUnico",
            "ConDatoAntes": len(valores_antes) - nulos_antes,
            "ConDatoDespues": int(
                df_limpio[columna].notna().sum()
            ),
            "NulosNuevos": nulos_despues - nulos_antes,
            "ValoresModificados": int(mascara_cambio.sum())
        })

        indices_cambio = valores_antes[
            mascara_cambio
        ].index[:10]

        for indice in indices_cambio:

            muestras_cambios.append({
                "Anio": anio,
                "Columna": columna,
                "FilaExcelFuente": df_limpio.loc[
                    indice,
                    "FilaExcelFuente"
                ],
                "ValorAntes": valores_antes.loc[indice],
                "ValorDespues": df_limpio.loc[
                    indice,
                    columna
                ]
            })

    # --------------------------------------------------------
    # CÓDIGOS MÚLTIPLES
    # --------------------------------------------------------

    for columna in columnas_codigo_multiple:

        valores_antes = df_limpio[columna].copy()

        df_limpio[columna] = (
            df_limpio[columna]
            .apply(
                lambda valor: limpiar_codigo(
                    valor,
                    conservar_varios=True,
                    minimo_digitos=5
                )
            )
            .astype("string")
        )

        mascara_cambio = (
            valores_antes.fillna("<NA>").astype(str)
            !=
            df_limpio[columna].fillna("<NA>").astype(str)
        )

        nulos_antes = int(
            valores_antes.apply(es_valor_vacio).sum()
        )

        nulos_despues = int(
            df_limpio[columna].isna().sum()
        )

        resumen_cambios.append({
            "Anio": anio,
            "Columna": columna,
            "TipoLimpieza": "CodigoMultiple",
            "ConDatoAntes": len(valores_antes) - nulos_antes,
            "ConDatoDespues": int(
                df_limpio[columna].notna().sum()
            ),
            "NulosNuevos": nulos_despues - nulos_antes,
            "ValoresModificados": int(mascara_cambio.sum())
        })

        indices_cambio = valores_antes[
            mascara_cambio
        ].index[:10]

        for indice in indices_cambio:

            muestras_cambios.append({
                "Anio": anio,
                "Columna": columna,
                "FilaExcelFuente": df_limpio.loc[
                    indice,
                    "FilaExcelFuente"
                ],
                "ValorAntes": valores_antes.loc[indice],
                "ValorDespues": df_limpio.loc[
                    indice,
                    columna
                ]
            })

    # --------------------------------------------------------
    # CAMPOS NUMÉRICOS
    # --------------------------------------------------------

    for columna in columnas_numericas:

        valores_antes = df_limpio[columna].copy()

        df_limpio[columna] = (
            df_limpio[columna]
            .apply(limpiar_numero)
        )

        df_limpio[columna] = pd.to_numeric(
            df_limpio[columna],
            errors="coerce"
        )

        mascara_cambio = (
            valores_antes.fillna("<NA>").astype(str)
            !=
            df_limpio[columna].fillna("<NA>").astype(str)
        )

        nulos_antes = int(
            valores_antes.apply(es_valor_vacio).sum()
        )

        nulos_despues = int(
            df_limpio[columna].isna().sum()
        )

        resumen_cambios.append({
            "Anio": anio,
            "Columna": columna,
            "TipoLimpieza": "Numerico",
            "ConDatoAntes": len(valores_antes) - nulos_antes,
            "ConDatoDespues": int(
                df_limpio[columna].notna().sum()
            ),
            "NulosNuevos": nulos_despues - nulos_antes,
            "ValoresModificados": int(mascara_cambio.sum())
        })

        # Solo guardar muestras que pasaron de dato a nulo
        mascara_nuevo_nulo = (
            ~valores_antes.apply(es_valor_vacio)
            &
            df_limpio[columna].isna()
        )

        indices_cambio = valores_antes[
            mascara_nuevo_nulo
        ].index[:10]

        for indice in indices_cambio:

            muestras_cambios.append({
                "Anio": anio,
                "Columna": columna,
                "FilaExcelFuente": df_limpio.loc[
                    indice,
                    "FilaExcelFuente"
                ],
                "ValorAntes": valores_antes.loc[indice],
                "ValorDespues": pd.NA
            })

    # --------------------------------------------------------
    # CAMPOS DE TEXTO
    # --------------------------------------------------------

    for columna in columnas_texto:

        valores_antes = df_limpio[columna].copy()

        df_limpio[columna] = (
            df_limpio[columna]
            .apply(limpiar_texto)
            .astype("string")
        )

        resumen_cambios.append({
            "Anio": anio,
            "Columna": columna,
            "TipoLimpieza": "Texto",
            "ConDatoAntes": int(
                ~valores_antes.apply(es_valor_vacio)
            .sum()
            ),
            "ConDatoDespues": int(
                df_limpio[columna].notna().sum()
            ),
            "NulosNuevos": 0
        })

    # --------------------------------------------------------
    # CAMPOS DE TRAZABILIDAD
    # --------------------------------------------------------

    df_limpio["AnioFuente"] = (
        pd.to_numeric(
            df_limpio["AnioFuente"],
            errors="coerce"
        )
        .astype("Int64")
    )

    df_limpio["FilaExcelFuente"] = (
        pd.to_numeric(
            df_limpio["FilaExcelFuente"],
            errors="coerce"
        )
        .astype("Int64")
    )

    return (
        df_limpio,
        resumen_cambios,
        muestras_cambios
    )


# ------------------------------------------------------------
# APLICAR LIMPIEZA A LOS CUATRO AÑOS
# ------------------------------------------------------------

datos_limpios_por_anio = {}
resumen_limpieza_total = []
muestras_limpieza_total = []
errores_limpieza = []

print("=" * 150)
print("APLICACIÓN DE LIMPIEZA POR AÑO - RELACIÓN CDP")
print("=" * 150)


for anio, df_homologado in (
    datos_homologados_por_anio.items()
):

    print("\n" + "=" * 150)
    print(f"AÑO: {anio}")
    print("=" * 150)

    try:

        (
            df_limpio,
            resumen_anio,
            muestras_anio
        ) = limpiar_dataframe_cdp(
            df_origen=df_homologado,
            anio=anio
        )

        datos_limpios_por_anio[anio] = (
            df_limpio.copy()
        )

        resumen_limpieza_total.extend(
            resumen_anio
        )

        muestras_limpieza_total.extend(
            muestras_anio
        )

        print(
            f"Registros antes: {len(df_homologado)}"
        )

        print(
            f"Registros después: {len(df_limpio)}"
        )

        print(
            f"Columnas antes: {len(df_homologado.columns)}"
        )

        print(
            f"Columnas después: {len(df_limpio.columns)}"
        )

        print("\nTIPOS DE DATOS RESULTANTES")

        print(
            df_limpio.dtypes.to_string()
        )

        print("\nPRIMEROS 3 REGISTROS LIMPIOS")

        columnas_muestra = [
            "FechaSolicitudCDP",
            "Proyecto",
            "Fondo",
            "PosicionPresupuestal",
            "NumeroContrato",
            "NumeroCDP",
            "NumeroRP",
            "ValorCDP",
            "SaldoCDP",
            "PlazoMeses",
            "PlazoDias",
            "FilaExcelFuente"
        ]

        print(
            df_limpio[
                columnas_muestra
            ]
            .head(3)
            .to_string(
                index=False,
                max_colwidth=70
            )
        )

    except Exception as error:

        print(f"ERROR: {error}")

        errores_limpieza.append({
            "Anio": anio,
            "Error": str(error)
        })


# ------------------------------------------------------------
# CREAR VARIABLES INDIVIDUALES LIMPIAS
# ------------------------------------------------------------

if len(datos_limpios_por_anio) == 4:

    df_cdp_2023_limpio = (
        datos_limpios_por_anio["2023"].copy()
    )

    df_cdp_2024_limpio = (
        datos_limpios_por_anio["2024"].copy()
    )

    df_cdp_2025_limpio = (
        datos_limpios_por_anio["2025"].copy()
    )

    df_cdp_2026_limpio = (
        datos_limpios_por_anio["2026"].copy()
    )


# ------------------------------------------------------------
# DATAFRAMES DE AUDITORÍA
# ------------------------------------------------------------

df_resumen_limpieza_cdp = pd.DataFrame(
    resumen_limpieza_total
)

df_muestras_limpieza_cdp = pd.DataFrame(
    muestras_limpieza_total
)


# ------------------------------------------------------------
# MOSTRAR NULOS NUEVOS
# ------------------------------------------------------------

print("\n" + "=" * 150)
print("VALORES CONVERTIDOS EN NULOS POR LAS REGLAS APROBADAS")
print("=" * 150)

if not df_resumen_limpieza_cdp.empty:

    nuevos_nulos = (
        df_resumen_limpieza_cdp[
            df_resumen_limpieza_cdp["NulosNuevos"] > 0
        ]
        .sort_values(
            by=["Anio", "Columna"]
        )
    )

    if nuevos_nulos.empty:

        print(
            "No se generaron nulos nuevos."
        )

    else:

        print(
            nuevos_nulos[
                [
                    "Anio",
                    "Columna",
                    "TipoLimpieza",
                    "ConDatoAntes",
                    "ConDatoDespues",
                    "NulosNuevos"
                ]
            ].to_string(index=False)
        )


# ------------------------------------------------------------
# MOSTRAR MUESTRAS DE CAMBIOS
# ------------------------------------------------------------

print("\n" + "=" * 150)
print("MUESTRAS DE VALORES MODIFICADOS")
print("=" * 150)

if df_muestras_limpieza_cdp.empty:

    print(
        "No se registraron valores modificados."
    )

else:

    print(
        df_muestras_limpieza_cdp
        .head(60)
        .to_string(
            index=False,
            max_colwidth=80
        )
    )


# ------------------------------------------------------------
# VALIDACIONES FINALES
# ------------------------------------------------------------

conteos_esperados = {
    "2023": 249,
    "2024": 313,
    "2025": 549,
    "2026": 231
}

conteos_correctos = (
    len(datos_limpios_por_anio) == 4
    and all(
        len(datos_limpios_por_anio[anio])
        == cantidad
        for anio, cantidad
        in conteos_esperados.items()
    )
)

estructuras_identicas = (
    len(datos_limpios_por_anio) == 4
    and all(
        list(df.columns)
        ==
        list(
            datos_limpios_por_anio["2023"].columns
        )
        for df in datos_limpios_por_anio.values()
    )
)

print("\n" + "=" * 150)

if errores_limpieza:

    print(
        "ESTADO: se presentaron errores durante "
        "la limpieza."
    )

    for error in errores_limpieza:
        print(
            f"- Año {error['Anio']}: "
            f"{error['Error']}"
        )

elif not conteos_correctos:

    print(
        "ADVERTENCIA: se alteró la cantidad de registros."
    )

elif not estructuras_identicas:

    print(
        "ADVERTENCIA: los cuatro años no conservaron "
        "la misma estructura."
    )

else:

    print(
        "ESTADO: la limpieza fue aplicada correctamente "
        "a copias separadas de los cuatro años. "
        "No se eliminaron registros y todavía no se "
        "unieron los DataFrames."
    )
```

```
# ============================================================
# PASO 19 - AUDITAR CÓDIGOS CONVERTIDOS A NULO
# GRUPO RELACION_CDP
# ============================================================

import pandas as pd


print("=" * 140)
print("AUDITORÍA DE CÓDIGOS CONVERTIDOS A NULO - RELACIÓN CDP")
print("=" * 140)


# ------------------------------------------------------------
# COLUMNAS DE CÓDIGO QUE DEBEMOS AUDITAR
# ------------------------------------------------------------

columnas_codigo_auditar = [
    "Proyecto",
    "Fondo",
    "PosicionPresupuestal",
    "NumeroEstudioPrevio",
    "SolicitudPedido",
    "NumeroContrato",
    "NumeroCDP",
    "NumeroRP"
]


# ------------------------------------------------------------
# CONTENEDOR DE RESULTADOS
# ------------------------------------------------------------

registros_codigo_nulo = []


# ------------------------------------------------------------
# COMPARAR HOMOLOGADO VS LIMPIO
# ------------------------------------------------------------

for anio in ["2023", "2024", "2025", "2026"]:

    df_antes = (
        datos_homologados_por_anio[anio]
        .copy()
        .reset_index(drop=True)
    )

    df_despues = (
        datos_limpios_por_anio[anio]
        .copy()
        .reset_index(drop=True)
    )

    print("\n" + "=" * 140)
    print(f"AÑO: {anio}")
    print("=" * 140)

    cantidad_anio = 0

    for columna in columnas_codigo_auditar:

        valor_antes_con_dato = (
            ~df_antes[columna].apply(es_valor_vacio)
        )

        valor_despues_nulo = (
            df_despues[columna].isna()
        )

        mascara = (
            valor_antes_con_dato
            &
            valor_despues_nulo
        )

        casos = df_antes.loc[
            mascara,
            [
                columna,
                "FechaSolicitudCDP",
                "Proyecto",
                "NumeroCDP",
                "ArchivoFuente",
                "FilaExcelFuente"
            ]
        ].copy()

        if casos.empty:
            continue

        cantidad_anio += len(casos)

        print(f"\nCOLUMNA: {columna}")
        print(f"Cantidad convertida a nulo: {len(casos)}")

        for indice in casos.index:

            valor_original = df_antes.loc[
                indice,
                columna
            ]

            print(
                f"\nFila Excel: "
                f"{df_antes.loc[indice, 'FilaExcelFuente']}"
            )

            print(
                f"Fecha: "
                f"{df_antes.loc[indice, 'FechaSolicitudCDP']}"
            )

            print(
                f"Proyecto: "
                f"{df_antes.loc[indice, 'Proyecto']}"
            )

            print(
                f"NumeroCDP: "
                f"{df_antes.loc[indice, 'NumeroCDP']}"
            )

            print(
                f"Valor original de {columna}: "
                f"{repr(valor_original)}"
            )

            registros_codigo_nulo.append({
                "Anio": anio,
                "Columna": columna,
                "FilaExcelFuente": df_antes.loc[
                    indice,
                    "FilaExcelFuente"
                ],
                "FechaSolicitudCDP": df_antes.loc[
                    indice,
                    "FechaSolicitudCDP"
                ],
                "Proyecto": df_antes.loc[
                    indice,
                    "Proyecto"
                ],
                "NumeroCDP": df_antes.loc[
                    indice,
                    "NumeroCDP"
                ],
                "ValorOriginal": valor_original
            })

    if cantidad_anio == 0:

        print(
            "No se encontraron códigos con contenido "
            "que hayan sido convertidos a nulo."
        )

    else:

        print(
            f"\nTotal de valores de código convertidos "
            f"a nulo en {anio}: {cantidad_anio}"
        )


# ------------------------------------------------------------
# DATAFRAME DE AUDITORÍA
# ------------------------------------------------------------

df_auditoria_codigos_nulos_cdp = pd.DataFrame(
    registros_codigo_nulo
)


# ------------------------------------------------------------
# RESUMEN GENERAL
# ------------------------------------------------------------

print("\n" + "=" * 140)
print("RESUMEN GENERAL")
print("=" * 140)

if df_auditoria_codigos_nulos_cdp.empty:

    print(
        "No hubo valores de código con contenido "
        "convertidos a nulo."
    )

else:

    resumen = (
        df_auditoria_codigos_nulos_cdp
        .groupby(
            ["Anio", "Columna"]
        )
        .size()
        .reset_index(
            name="Cantidad"
        )
    )

    print(
        resumen.to_string(
            index=False
        )
    )

    print(
        "\nTotal general de valores auditados: "
        f"{len(df_auditoria_codigos_nulos_cdp)}"
    )


print("\n" + "=" * 140)
print(
    "ESTADO: auditoría terminada. "
    "No se modificaron ni unieron los datos."
)
```

```
# ============================================================
# PASO 20 - CONSOLIDAR LOS CUATRO AÑOS
# GRUPO RELACION_CDP
# ============================================================

import pandas as pd


print("=" * 150)
print("CONSOLIDACIÓN DE LOS CUATRO AÑOS - RELACIÓN CDP")
print("=" * 150)


# ------------------------------------------------------------
# VALIDAR QUE LOS CUATRO DATAFRAMES LIMPIOS EXISTAN
# ------------------------------------------------------------

anios_esperados = [
    "2023",
    "2024",
    "2025",
    "2026"
]

anios_disponibles = sorted(
    datos_limpios_por_anio.keys()
)

if anios_disponibles != anios_esperados:

    raise ValueError(
        "No están disponibles los cuatro años esperados. "
        f"Años encontrados: {anios_disponibles}"
    )


# ------------------------------------------------------------
# GUARDAR CONTEOS ANTES DE LA UNIÓN
# ------------------------------------------------------------

conteos_antes_union = {
    anio: len(datos_limpios_por_anio[anio])
    for anio in anios_esperados
}

print("\nREGISTROS ANTES DE LA CONSOLIDACIÓN")

for anio, cantidad in conteos_antes_union.items():
    print(f"Año {anio}: {cantidad} registros")

total_esperado = sum(
    conteos_antes_union.values()
)

print(f"\nTotal esperado: {total_esperado} registros")


# ------------------------------------------------------------
# VALIDAR QUE LA ESTRUCTURA SEA IDÉNTICA
# ------------------------------------------------------------

columnas_base = list(
    datos_limpios_por_anio["2023"].columns
)

estructura_identica = all(
    list(datos_limpios_por_anio[anio].columns)
    == columnas_base
    for anio in anios_esperados
)

if not estructura_identica:

    raise ValueError(
        "Los cuatro DataFrames no tienen la misma estructura."
    )


# ------------------------------------------------------------
# CONSOLIDAR
# ------------------------------------------------------------

df_cdp_consolidado = pd.concat(
    [
        datos_limpios_por_anio["2023"],
        datos_limpios_por_anio["2024"],
        datos_limpios_por_anio["2025"],
        datos_limpios_por_anio["2026"]
    ],
    axis=0,
    ignore_index=True
)


# ------------------------------------------------------------
# ORDENAR SIN ALTERAR LA GRANULARIDAD
# ------------------------------------------------------------

df_cdp_consolidado = (
    df_cdp_consolidado
    .sort_values(
        by=[
            "FechaSolicitudCDP",
            "AnioFuente",
            "FilaExcelFuente"
        ],
        na_position="last"
    )
    .reset_index(drop=True)
)


# ------------------------------------------------------------
# CREAR IDENTIFICADOR TÉCNICO
# ------------------------------------------------------------

df_cdp_consolidado.insert(
    0,
    "IdRegistro",
    range(
        1,
        len(df_cdp_consolidado) + 1
    )
)

df_cdp_consolidado["IdRegistro"] = (
    df_cdp_consolidado["IdRegistro"]
    .astype("Int64")
)


# ------------------------------------------------------------
# VALIDAR CONTEOS POR AÑO
# ------------------------------------------------------------

conteos_despues_union = (
    df_cdp_consolidado
    .groupby(
        "AnioFuente",
        dropna=False
    )
    .size()
    .reset_index(
        name="CantidadRegistros"
    )
)

print("\n" + "=" * 150)
print("REGISTROS DESPUÉS DE LA CONSOLIDACIÓN")
print("=" * 150)

print(
    conteos_despues_union.to_string(
        index=False
    )
)


# ------------------------------------------------------------
# VALIDAR FECHAS FRENTE AL AÑO FUENTE
# ------------------------------------------------------------

anio_fecha = (
    df_cdp_consolidado["FechaSolicitudCDP"]
    .dt.year
    .astype("Int64")
)

mascara_fecha_otro_anio = (
    df_cdp_consolidado["FechaSolicitudCDP"].notna()
    &
    (
        anio_fecha
        != df_cdp_consolidado["AnioFuente"]
    )
)

registros_fecha_otro_anio = (
    df_cdp_consolidado.loc[
        mascara_fecha_otro_anio,
        [
            "IdRegistro",
            "FechaSolicitudCDP",
            "AnioFuente",
            "Proyecto",
            "NumeroCDP",
            "ArchivoFuente",
            "FilaExcelFuente"
        ]
    ]
)


# ------------------------------------------------------------
# IDENTIFICAR DUPLICADOS EXACTOS DE NEGOCIO
# NO SE ELIMINAN
# ------------------------------------------------------------

columnas_duplicado_exacto = [
    columna
    for columna in df_cdp_consolidado.columns
    if columna not in [
        "IdRegistro",
        "ArchivoFuente",
        "FilaExcelFuente"
    ]
]

mascara_duplicado_exacto = (
    df_cdp_consolidado
    .duplicated(
        subset=columnas_duplicado_exacto,
        keep=False
    )
)

df_duplicados_exactos_cdp = (
    df_cdp_consolidado.loc[
        mascara_duplicado_exacto
    ]
    .copy()
)


# ------------------------------------------------------------
# IDENTIFICAR REPETICIONES DE CDP
# NO SIGNIFICAN NECESARIAMENTE DUPLICADOS
# ------------------------------------------------------------

conteo_por_cdp = (
    df_cdp_consolidado[
        df_cdp_consolidado["NumeroCDP"].notna()
    ]
    .groupby(
        [
            "AnioFuente",
            "NumeroCDP"
        ],
        dropna=False
    )
    .size()
    .reset_index(
        name="CantidadFilas"
    )
)

df_cdp_repetidos = (
    conteo_por_cdp[
        conteo_por_cdp["CantidadFilas"] > 1
    ]
    .sort_values(
        by=[
            "AnioFuente",
            "CantidadFilas",
            "NumeroCDP"
        ],
        ascending=[
            True,
            False,
            True
        ]
    )
)


# ------------------------------------------------------------
# MOSTRAR PRIMEROS Y ÚLTIMOS REGISTROS
# ------------------------------------------------------------

columnas_muestra = [
    "IdRegistro",
    "FechaSolicitudCDP",
    "Proyecto",
    "NumeroCDP",
    "NumeroContrato",
    "NumeroRP",
    "ValorCDP",
    "AnioFuente",
    "FilaExcelFuente"
]

print("\n" + "=" * 150)
print("PRIMEROS 5 REGISTROS DEL CONSOLIDADO")
print("=" * 150)

print(
    df_cdp_consolidado[
        columnas_muestra
    ]
    .head(5)
    .to_string(
        index=False,
        max_colwidth=60
    )
)

print("\n" + "=" * 150)
print("ÚLTIMOS 5 REGISTROS DEL CONSOLIDADO")
print("=" * 150)

print(
    df_cdp_consolidado[
        columnas_muestra
    ]
    .tail(5)
    .to_string(
        index=False,
        max_colwidth=60
    )
)


# ------------------------------------------------------------
# RESUMEN DE CALIDAD
# ------------------------------------------------------------

print("\n" + "=" * 150)
print("RESUMEN DE VALIDACIONES")
print("=" * 150)

print(
    f"Total esperado: "
    f"{total_esperado}"
)

print(
    f"Total consolidado: "
    f"{len(df_cdp_consolidado)}"
)

print(
    f"Cantidad de columnas: "
    f"{len(df_cdp_consolidado.columns)}"
)

print(
    f"Fechas que no corresponden al año fuente: "
    f"{len(registros_fecha_otro_anio)}"
)

print(
    f"Filas involucradas en duplicados exactos: "
    f"{len(df_duplicados_exactos_cdp)}"
)

print(
    f"Combinaciones Año-CDP repetidas: "
    f"{len(df_cdp_repetidos)}"
)


# ------------------------------------------------------------
# MOSTRAR DUPLICADOS EXACTOS SI EXISTEN
# ------------------------------------------------------------

print("\n" + "=" * 150)
print("DUPLICADOS EXACTOS DETECTADOS")
print("=" * 150)

if df_duplicados_exactos_cdp.empty:

    print(
        "No se encontraron duplicados exactos."
    )

else:

    print(
        df_duplicados_exactos_cdp[
            columnas_muestra
        ]
        .head(30)
        .to_string(
            index=False,
            max_colwidth=60
        )
    )


# ------------------------------------------------------------
# MOSTRAR CDP REPETIDOS
# ------------------------------------------------------------

print("\n" + "=" * 150)
print("CDP REPETIDOS DENTRO DEL MISMO AÑO")
print("=" * 150)

if df_cdp_repetidos.empty:

    print(
        "No se encontraron CDP repetidos."
    )

else:

    print(
        df_cdp_repetidos
        .head(40)
        .to_string(
            index=False
        )
    )


# ------------------------------------------------------------
# VALIDACIÓN FINAL
# ------------------------------------------------------------

conteo_correcto = (
    len(df_cdp_consolidado)
    == total_esperado
    == 1342
)

anios_correctos = (
    set(
        df_cdp_consolidado[
            "AnioFuente"
        ]
        .dropna()
        .astype(int)
        .tolist()
    )
    == {
        2023,
        2024,
        2025,
        2026
    }
)

ids_unicos = (
    df_cdp_consolidado[
        "IdRegistro"
    ]
    .is_unique
)

print("\n" + "=" * 150)

if not conteo_correcto:

    print(
        "ADVERTENCIA: el total consolidado "
        "no coincide con los 1.342 registros esperados."
    )

elif not anios_correctos:

    print(
        "ADVERTENCIA: los años del consolidado "
        "no corresponden a 2023, 2024, 2025 y 2026."
    )

elif not ids_unicos:

    print(
        "ADVERTENCIA: IdRegistro contiene valores repetidos."
    )

else:

    print(
        "ESTADO: los cuatro años fueron consolidados "
        "correctamente en 1.342 registros. "
        "No se eliminaron duplicados ni se exportaron archivos."
    )
```

```
# ============================================================
# MAESTRA DE PROYECTOS - PASO 3
# VALIDAR LLAVES, DUPLICADOS Y COBERTURA
# ============================================================

import pandas as pd
import numpy as np
import re
import math


print("=" * 130)
print("VALIDACIÓN DE LLAVES Y COBERTURA - MAESTRA DE PROYECTOS")
print("=" * 130)


# ------------------------------------------------------------
# FUNCIÓN PARA LIMPIAR CÓDIGOS DE PROYECTO
# ------------------------------------------------------------

def limpiar_codigo_proyecto(valor):

    if valor is None or pd.isna(valor):
        return pd.NA

    if isinstance(valor, (int, np.integer)):
        return str(valor)

    if isinstance(valor, (float, np.floating)):

        if math.isnan(valor):
            return pd.NA

        if valor.is_integer():
            return str(int(valor))

    texto = str(valor).strip()

    coincidencia = re.search(
        r"(?<!\d)\d{5,}(?!\d)",
        texto
    )

    if not coincidencia:
        return pd.NA

    return coincidencia.group(0)


# ------------------------------------------------------------
# LEER LAS HOJAS NECESARIAS
# ------------------------------------------------------------

tablas_maestra = {}

for anio in ["2023", "2024"]:

    ruta = config_maestra[anio]["archivo"]

    proyectos = pd.read_excel(
        ruta,
        sheet_name="PROYECTOS",
        header=0,
        engine="openpyxl"
    )

    subsecretarias = pd.read_excel(
        ruta,
        sheet_name="SUBSECRETARÍA",
        header=0,
        engine="openpyxl"
    )

    responsables = pd.read_excel(
        ruta,
        sheet_name="RESPONSABLES",
        header=0,
        engine="openpyxl"
    )

    tablas_maestra[anio] = {
        "proyectos": proyectos,
        "subsecretarias": subsecretarias,
        "responsables": responsables
    }


for anio in ["2025", "2026"]:

    ruta = config_maestra[anio]["archivo"]

    proyectos = pd.read_excel(
        ruta,
        sheet_name="Proyectos",
        header=0,
        engine="openpyxl"
    )

    tablas_maestra[anio] = {
        "proyectos": proyectos
    }


# ------------------------------------------------------------
# VALIDAR LAS 3 COLUMNAS PROYECTO EN 2023 Y 2024
# ------------------------------------------------------------

print("\n" + "=" * 130)
print("VALIDACIÓN DE COLUMNAS PROYECTO, PROYECTO.1 Y PROYECTO.2")
print("=" * 130)

for anio in ["2023", "2024"]:

    df = tablas_maestra[anio]["proyectos"].copy()

    proyecto_1 = df["Proyecto"].apply(
        limpiar_codigo_proyecto
    )

    proyecto_2 = df["Proyecto.1"].apply(
        limpiar_codigo_proyecto
    )

    proyecto_3 = df["Proyecto.2"].apply(
        limpiar_codigo_proyecto
    )

    diferencias_1_2 = int(
        (
            proyecto_1.fillna("<NA>")
            != proyecto_2.fillna("<NA>")
        ).sum()
    )

    diferencias_1_3 = int(
        (
            proyecto_1.fillna("<NA>")
            != proyecto_3.fillna("<NA>")
        ).sum()
    )

    print(
        f"Año {anio}: "
        f"Proyecto vs Proyecto.1 = {diferencias_1_2} diferencias | "
        f"Proyecto vs Proyecto.2 = {diferencias_1_3} diferencias"
    )


# ------------------------------------------------------------
# PREPARAR TABLAS ESTANDARIZADAS
# ------------------------------------------------------------

maestras_proyectos_por_anio = {}

for anio in ["2023", "2024"]:

    df = tablas_maestra[anio]["proyectos"].copy()

    df_estandar = pd.DataFrame({
        "Proyecto": df["Proyecto"].apply(
            limpiar_codigo_proyecto
        ),
        "DescripcionProyecto": df["Denominación"]
    })

    df_estandar = (
        df_estandar
        .dropna(subset=["Proyecto"])
        .drop_duplicates()
        .reset_index(drop=True)
    )

    maestras_proyectos_por_anio[anio] = df_estandar


for anio in ["2025", "2026"]:

    df = tablas_maestra[anio]["proyectos"].copy()

    df_estandar = pd.DataFrame({
        "Proyecto": df["Programa/ proyecto"].apply(
            limpiar_codigo_proyecto
        ),
        "DescripcionProyecto": df[
            "SECCIÓN/OBJETO DE GASTO"
        ]
    })

    df_estandar = (
        df_estandar
        .dropna(subset=["Proyecto"])
        .drop_duplicates()
        .reset_index(drop=True)
    )

    maestras_proyectos_por_anio[anio] = df_estandar


# ------------------------------------------------------------
# PREPARAR SUBSECRETARÍAS Y RESPONSABLES
# ------------------------------------------------------------

auxiliares_por_anio = {}

for anio in ["2023", "2024"]:

    df_sub = tablas_maestra[anio][
        "subsecretarias"
    ].copy()

    df_resp = tablas_maestra[anio][
        "responsables"
    ].copy()

    df_sub_estandar = pd.DataFrame({
        "Proyecto": df_sub["CÓDIGO NUEVO"].apply(
            limpiar_codigo_proyecto
        ),
        "Subsecretaria": df_sub["SUBDEPENDENCIA"]
    })

    df_resp_estandar = pd.DataFrame({
        "Proyecto": df_resp["Código"].apply(
            limpiar_codigo_proyecto
        ),
        "Responsable": df_resp["Responsable"]
    })

    auxiliares_por_anio[anio] = {
        "subsecretarias": (
            df_sub_estandar
            .dropna(subset=["Proyecto"])
            .drop_duplicates()
            .reset_index(drop=True)
        ),
        "responsables": (
            df_resp_estandar
            .dropna(subset=["Proyecto"])
            .drop_duplicates()
            .reset_index(drop=True)
        )
    }


# ------------------------------------------------------------
# VALIDAR DUPLICADOS POR PROYECTO
# ------------------------------------------------------------

print("\n" + "=" * 130)
print("DUPLICADOS POR CÓDIGO DE PROYECTO")
print("=" * 130)

for anio, df in maestras_proyectos_por_anio.items():

    duplicados = (
        df.groupby("Proyecto")
        .size()
        .reset_index(name="Cantidad")
    )

    duplicados = duplicados[
        duplicados["Cantidad"] > 1
    ]

    print(
        f"Año {anio}: "
        f"{len(df)} filas únicas Proyecto-Descripción | "
        f"{len(duplicados)} códigos con más de una descripción"
    )


# ------------------------------------------------------------
# COMPARAR 2023 VS 2024
# ------------------------------------------------------------

print("\n" + "=" * 130)
print("COMPARACIÓN DE TABLAS 2023 Y 2024")
print("=" * 130)

proyectos_2023 = set(
    maestras_proyectos_por_anio["2023"]["Proyecto"]
)

proyectos_2024 = set(
    maestras_proyectos_por_anio["2024"]["Proyecto"]
)

print(
    f"Proyectos comunes: "
    f"{len(proyectos_2023 & proyectos_2024)}"
)

print(
    f"Solo en 2023: "
    f"{len(proyectos_2023 - proyectos_2024)}"
)

print(
    f"Solo en 2024: "
    f"{len(proyectos_2024 - proyectos_2023)}"
)


# ------------------------------------------------------------
# COBERTURA DE AUXILIARES SOBRE LOS PROYECTOS UTILIZADOS EN CDP
# ------------------------------------------------------------

print("\n" + "=" * 130)
print("COBERTURA DE SUBSECRETARÍA Y RESPONSABLE")
print("=" * 130)

proyectos_usados_cdp = (
    df_cdp_consolidado[
        [
            "Proyecto",
            "AnioFuente"
        ]
    ]
    .dropna(subset=["Proyecto"])
    .drop_duplicates()
)

for anio in ["2023", "2024", "2025", "2026"]:

    proyectos_anio = set(
        proyectos_usados_cdp.loc[
            proyectos_usados_cdp["AnioFuente"]
            == int(anio),
            "Proyecto"
        ].astype(str)
    )

    # Para 2025 y 2026 se evalúa cobertura usando
    # los auxiliares históricos de 2024
    anio_auxiliar = (
        anio
        if anio in ["2023", "2024"]
        else "2024"
    )

    proyectos_sub = set(
        auxiliares_por_anio[anio_auxiliar][
            "subsecretarias"
        ]["Proyecto"]
    )

    proyectos_resp = set(
        auxiliares_por_anio[anio_auxiliar][
            "responsables"
        ]["Proyecto"]
    )

    con_subsecretaria = (
        proyectos_anio & proyectos_sub
    )

    con_responsable = (
        proyectos_anio & proyectos_resp
    )

    print(f"\nAño CDP {anio}:")
    print(
        f"  Proyectos utilizados: "
        f"{len(proyectos_anio)}"
    )
    print(
        f"  Con subsecretaría: "
        f"{len(con_subsecretaria)}"
    )
    print(
        f"  Sin subsecretaría: "
        f"{len(proyectos_anio - proyectos_sub)}"
    )
    print(
        f"  Con responsable: "
        f"{len(con_responsable)}"
    )
    print(
        f"  Sin responsable: "
        f"{len(proyectos_anio - proyectos_resp)}"
    )

    if proyectos_anio - proyectos_resp:

        print(
            "  Proyectos sin responsable: "
            + ", ".join(
                sorted(
                    proyectos_anio
                    - proyectos_resp
                )
            )
        )


print("\n" + "=" * 130)
print(
    "ESTADO: validación terminada. "
    "Todavía no se construyó ni exportó la maestra."
)
```

```
# ============================================================
# MAESTRA DE PROYECTOS - PASO 4 AJUSTADO
# CREAR:
# 1. MAESTRA PARA RELACIÓN CDP
# 2. HISTÓRICO GENERAL DE PROYECTOS
# ============================================================

import pandas as pd


print("=" * 140)
print("CONSTRUCCIÓN DE MAESTRA CDP E HISTÓRICO DE PROYECTOS")
print("=" * 140)


# ============================================================
# 1. HISTÓRICO GENERAL DE PROYECTOS
# ============================================================

# Se usa 2024 como fuente principal porque 2023 y 2024
# contienen los mismos 3.001 códigos de proyecto.

df_historico_proyectos = (
    maestras_proyectos_por_anio["2024"]
    .copy()
)

df_historico_proyectos["AnioFuenteCatalogo"] = 2024
df_historico_proyectos["ArchivoFuenteCatalogo"] = (
    "1.2. RELACION CDP A DIC 27 DE 2024.xlsx"
)

# Agregar subsecretaría y responsable cuando exista coincidencia exacta

df_subsecretarias_historico = (
    auxiliares_por_anio["2024"][
        "subsecretarias"
    ]
    .copy()
)

df_responsables_historico = (
    auxiliares_por_anio["2024"][
        "responsables"
    ]
    .copy()
)

df_historico_proyectos = (
    df_historico_proyectos
    .merge(
        df_subsecretarias_historico,
        on="Proyecto",
        how="left",
        validate="many_to_one"
    )
    .merge(
        df_responsables_historico,
        on="Proyecto",
        how="left",
        validate="many_to_one"
    )
)


# ------------------------------------------------------------
# Identificar códigos con más de una descripción
# ------------------------------------------------------------

conteo_descripciones_historico = (
    df_historico_proyectos
    .groupby("Proyecto")[
        "DescripcionProyecto"
    ]
    .nunique(dropna=True)
    .reset_index(
        name="CantidadDescripciones"
    )
)

df_historico_proyectos = (
    df_historico_proyectos
    .merge(
        conteo_descripciones_historico,
        on="Proyecto",
        how="left"
    )
)

df_historico_proyectos[
    "TieneMultiplesDescripciones"
] = (
    df_historico_proyectos[
        "CantidadDescripciones"
    ] > 1
)


# ============================================================
# 2. MAESTRA DE PROYECTOS UTILIZADOS EN RELACIÓN CDP
# ============================================================

df_proyectos_usados_cdp = (
    df_cdp_consolidado[
        [
            "AnioFuente",
            "Proyecto",
            "DescripcionProyecto"
        ]
    ]
    .dropna(subset=["Proyecto"])
    .copy()
)


# ------------------------------------------------------------
# Elegir una descripción representativa por año y proyecto
# ------------------------------------------------------------

def descripcion_mas_frecuente(serie):

    valores = (
        serie
        .dropna()
        .astype(str)
        .str.strip()
    )

    valores = valores[
        valores != ""
    ]

    if valores.empty:
        return pd.NA

    return valores.value_counts().index[0]


df_maestra_proyectos_cdp = (
    df_proyectos_usados_cdp
    .groupby(
        [
            "AnioFuente",
            "Proyecto"
        ],
        as_index=False
    )
    .agg(
        DescripcionProyecto=(
            "DescripcionProyecto",
            descripcion_mas_frecuente
        )
    )
)


# ------------------------------------------------------------
# Agregar cantidad de descripciones observadas
# ------------------------------------------------------------

conteo_descripciones_cdp = (
    df_proyectos_usados_cdp
    .groupby(
        [
            "AnioFuente",
            "Proyecto"
        ]
    )[
        "DescripcionProyecto"
    ]
    .nunique(dropna=True)
    .reset_index(
        name="CantidadDescripciones"
    )
)

df_maestra_proyectos_cdp = (
    df_maestra_proyectos_cdp
    .merge(
        conteo_descripciones_cdp,
        on=[
            "AnioFuente",
            "Proyecto"
        ],
        how="left"
    )
)

df_maestra_proyectos_cdp[
    "TieneMultiplesDescripciones"
] = (
    df_maestra_proyectos_cdp[
        "CantidadDescripciones"
    ] > 1
)


# ------------------------------------------------------------
# Enriquecer desde el histórico por coincidencia exacta
# ------------------------------------------------------------

df_catalogo_enriquecimiento = (
    df_historico_proyectos[
        [
            "Proyecto",
            "Subsecretaria",
            "Responsable"
        ]
    ]
    .drop_duplicates(
        subset=["Proyecto"]
    )
)

df_maestra_proyectos_cdp = (
    df_maestra_proyectos_cdp
    .merge(
        df_catalogo_enriquecimiento,
        on="Proyecto",
        how="left",
        validate="many_to_one"
    )
)


# ------------------------------------------------------------
# Indicadores de cobertura
# ------------------------------------------------------------

df_maestra_proyectos_cdp[
    "TieneSubsecretaria"
] = (
    df_maestra_proyectos_cdp[
        "Subsecretaria"
    ]
    .notna()
)

df_maestra_proyectos_cdp[
    "TieneResponsable"
] = (
    df_maestra_proyectos_cdp[
        "Responsable"
    ]
    .notna()
)


# ------------------------------------------------------------
# Ordenar resultados
# ------------------------------------------------------------

df_historico_proyectos = (
    df_historico_proyectos
    .sort_values(
        by=[
            "Proyecto",
            "DescripcionProyecto"
        ]
    )
    .reset_index(drop=True)
)

df_maestra_proyectos_cdp = (
    df_maestra_proyectos_cdp
    .sort_values(
        by=[
            "AnioFuente",
            "Proyecto"
        ]
    )
    .reset_index(drop=True)
)


# ============================================================
# 3. VALIDACIONES
# ============================================================

print("\nHISTÓRICO GENERAL DE PROYECTOS")
print("-" * 140)

print(
    f"Filas Proyecto-Descripción: "
    f"{len(df_historico_proyectos)}"
)

print(
    f"Códigos únicos de proyecto: "
    f"{df_historico_proyectos['Proyecto'].nunique()}"
)

print(
    f"Códigos con múltiples descripciones: "
    f"{df_historico_proyectos.loc[
        df_historico_proyectos[
            'TieneMultiplesDescripciones'
        ],
        'Proyecto'
    ].nunique()}"
)


print("\nMAESTRA DE PROYECTOS PARA RELACIÓN CDP")
print("-" * 140)

print(
    f"Filas de la maestra: "
    f"{len(df_maestra_proyectos_cdp)}"
)

print(
    f"Proyectos-año únicos: "
    f"{df_maestra_proyectos_cdp[
        ['AnioFuente', 'Proyecto']
    ].drop_duplicates().shape[0]}"
)


# ------------------------------------------------------------
# Cobertura por año
# ------------------------------------------------------------

resumen_cobertura = (
    df_maestra_proyectos_cdp
    .groupby("AnioFuente")
    .agg(
        CantidadProyectos=(
            "Proyecto",
            "count"
        ),
        ConSubsecretaria=(
            "TieneSubsecretaria",
            "sum"
        ),
        ConResponsable=(
            "TieneResponsable",
            "sum"
        ),
        ConMultiplesDescripciones=(
            "TieneMultiplesDescripciones",
            "sum"
        )
    )
    .reset_index()
)

print("\nCOBERTURA DE LA MAESTRA CDP")

print(
    resumen_cobertura.to_string(
        index=False
    )
)


# ------------------------------------------------------------
# Mostrar proyectos históricos con varias descripciones
# ------------------------------------------------------------

print("\n" + "=" * 140)
print("PROYECTOS HISTÓRICOS CON MÚLTIPLES DESCRIPCIONES")
print("=" * 140)

df_historico_multiples_descripciones = (
    df_historico_proyectos[
        df_historico_proyectos[
            "TieneMultiplesDescripciones"
        ]
    ]
    .copy()
)

if df_historico_multiples_descripciones.empty:

    print(
        "No se encontraron proyectos con varias descripciones."
    )

else:

    print(
        df_historico_multiples_descripciones[
            [
                "Proyecto",
                "DescripcionProyecto",
                "CantidadDescripciones"
            ]
        ].to_string(
            index=False,
            max_colwidth=100
        )
    )


print("\n" + "=" * 140)
print(
    "ESTADO: se crearon dos tablas provisionales: "
    "la maestra de proyectos usados en RELACION_CDP "
    "y el histórico general de proyectos. "
    "Todavía no se exportaron archivos."
)
```

```
# ============================================================
# MAESTRA DE PROYECTOS - PASO 4 CORREGIDO
# CREAR:
# 1. MAESTRA PARA RELACIÓN CDP
# 2. HISTÓRICO GENERAL DE PROYECTOS
# AGRUPANDO RESPONSABLES MÚLTIPLES EN UNA SOLA CELDA
# ============================================================

import pandas as pd


print("=" * 140)
print("CONSTRUCCIÓN DE MAESTRA CDP E HISTÓRICO DE PROYECTOS")
print("=" * 140)


# ------------------------------------------------------------
# FUNCIÓN PARA AGRUPAR TEXTOS DISTINTOS
# ------------------------------------------------------------

def agrupar_textos_unicos(serie, separador="/"):

    valores = (
        serie
        .dropna()
        .astype(str)
        .str.strip()
    )

    valores = valores[
        valores != ""
    ]

    if valores.empty:
        return pd.NA

    valores_unicos = list(
        dict.fromkeys(valores)
    )

    return separador.join(valores_unicos)


# ============================================================
# 1. PREPARAR AUXILIARES SIN DUPLICAR PROYECTOS
# ============================================================

df_subsecretarias_historico = (
    auxiliares_por_anio["2024"]["subsecretarias"]
    .copy()
)

df_responsables_historico = (
    auxiliares_por_anio["2024"]["responsables"]
    .copy()
)


# ------------------------------------------------------------
# AGRUPAR SUBSECRETARÍAS POR PROYECTO
# ------------------------------------------------------------

df_subsecretarias_historico = (
    df_subsecretarias_historico
    .groupby(
        "Proyecto",
        as_index=False
    )
    .agg(
        Subsecretaria=(
            "Subsecretaria",
            agrupar_textos_unicos
        )
    )
)


# ------------------------------------------------------------
# AGRUPAR RESPONSABLES POR PROYECTO
# ------------------------------------------------------------

df_responsables_historico = (
    df_responsables_historico
    .groupby(
        "Proyecto",
        as_index=False
    )
    .agg(
        Responsable=(
            "Responsable",
            agrupar_textos_unicos
        )
    )
)


# ------------------------------------------------------------
# VALIDAR UNICIDAD DE AUXILIARES
# ------------------------------------------------------------

subsecretarias_unicas = (
    df_subsecretarias_historico["Proyecto"]
    .is_unique
)

responsables_unicos = (
    df_responsables_historico["Proyecto"]
    .is_unique
)

print(
    f"Subsecretarías con llave única: "
    f"{subsecretarias_unicas}"
)

print(
    f"Responsables con llave única: "
    f"{responsables_unicos}"
)


# ============================================================
# 2. HISTÓRICO GENERAL DE PROYECTOS
# ============================================================

df_historico_proyectos = (
    maestras_proyectos_por_anio["2024"]
    .copy()
)

df_historico_proyectos["AnioFuenteCatalogo"] = 2024

df_historico_proyectos["ArchivoFuenteCatalogo"] = (
    "1.2. RELACION CDP A DIC 27 DE 2024.xlsx"
)


# ------------------------------------------------------------
# ENRIQUECER HISTÓRICO
# ------------------------------------------------------------

df_historico_proyectos = (
    df_historico_proyectos
    .merge(
        df_subsecretarias_historico,
        on="Proyecto",
        how="left",
        validate="many_to_one"
    )
    .merge(
        df_responsables_historico,
        on="Proyecto",
        how="left",
        validate="many_to_one"
    )
)


# ------------------------------------------------------------
# IDENTIFICAR MÚLTIPLES DESCRIPCIONES
# ------------------------------------------------------------

conteo_descripciones_historico = (
    df_historico_proyectos
    .groupby("Proyecto")[
        "DescripcionProyecto"
    ]
    .nunique(dropna=True)
    .reset_index(
        name="CantidadDescripciones"
    )
)

df_historico_proyectos = (
    df_historico_proyectos
    .merge(
        conteo_descripciones_historico,
        on="Proyecto",
        how="left",
        validate="many_to_one"
    )
)

df_historico_proyectos[
    "TieneMultiplesDescripciones"
] = (
    df_historico_proyectos[
        "CantidadDescripciones"
    ] > 1
)


# ============================================================
# 3. MAESTRA DE PROYECTOS UTILIZADOS EN RELACIÓN CDP
# ============================================================

df_proyectos_usados_cdp = (
    df_cdp_consolidado[
        [
            "AnioFuente",
            "Proyecto",
            "DescripcionProyecto"
        ]
    ]
    .dropna(subset=["Proyecto"])
    .copy()
)


# ------------------------------------------------------------
# ELEGIR DESCRIPCIÓN MÁS FRECUENTE
# ------------------------------------------------------------

def descripcion_mas_frecuente(serie):

    valores = (
        serie
        .dropna()
        .astype(str)
        .str.strip()
    )

    valores = valores[
        valores != ""
    ]

    if valores.empty:
        return pd.NA

    return valores.value_counts().index[0]


df_maestra_proyectos_cdp = (
    df_proyectos_usados_cdp
    .groupby(
        [
            "AnioFuente",
            "Proyecto"
        ],
        as_index=False
    )
    .agg(
        DescripcionProyecto=(
            "DescripcionProyecto",
            descripcion_mas_frecuente
        )
    )
)


# ------------------------------------------------------------
# CONTAR DESCRIPCIONES OBSERVADAS
# ------------------------------------------------------------

conteo_descripciones_cdp = (
    df_proyectos_usados_cdp
    .groupby(
        [
            "AnioFuente",
            "Proyecto"
        ]
    )[
        "DescripcionProyecto"
    ]
    .nunique(dropna=True)
    .reset_index(
        name="CantidadDescripciones"
    )
)

df_maestra_proyectos_cdp = (
    df_maestra_proyectos_cdp
    .merge(
        conteo_descripciones_cdp,
        on=[
            "AnioFuente",
            "Proyecto"
        ],
        how="left",
        validate="one_to_one"
    )
)

df_maestra_proyectos_cdp[
    "TieneMultiplesDescripciones"
] = (
    df_maestra_proyectos_cdp[
        "CantidadDescripciones"
    ] > 1
)


# ------------------------------------------------------------
# ENRIQUECER MAESTRA CDP
# ------------------------------------------------------------

df_maestra_proyectos_cdp = (
    df_maestra_proyectos_cdp
    .merge(
        df_subsecretarias_historico,
        on="Proyecto",
        how="left",
        validate="many_to_one"
    )
    .merge(
        df_responsables_historico,
        on="Proyecto",
        how="left",
        validate="many_to_one"
    )
)


# ------------------------------------------------------------
# INDICADORES DE COBERTURA
# ------------------------------------------------------------

df_maestra_proyectos_cdp[
    "TieneSubsecretaria"
] = (
    df_maestra_proyectos_cdp[
        "Subsecretaria"
    ]
    .notna()
)

df_maestra_proyectos_cdp[
    "TieneResponsable"
] = (
    df_maestra_proyectos_cdp[
        "Responsable"
    ]
    .notna()
)


# ------------------------------------------------------------
# ORDENAR
# ------------------------------------------------------------

df_historico_proyectos = (
    df_historico_proyectos
    .sort_values(
        [
            "Proyecto",
            "DescripcionProyecto"
        ]
    )
    .reset_index(drop=True)
)

df_maestra_proyectos_cdp = (
    df_maestra_proyectos_cdp
    .sort_values(
        [
            "AnioFuente",
            "Proyecto"
        ]
    )
    .reset_index(drop=True)
)


# ============================================================
# 4. VALIDACIONES
# ============================================================

print("\nHISTÓRICO GENERAL DE PROYECTOS")
print("-" * 140)

print(
    f"Filas Proyecto-Descripción: "
    f"{len(df_historico_proyectos)}"
)

print(
    f"Códigos únicos: "
    f"{df_historico_proyectos['Proyecto'].nunique()}"
)

print(
    f"Códigos con múltiples descripciones: "
    f"{df_historico_proyectos.loc[
        df_historico_proyectos[
            'TieneMultiplesDescripciones'
        ],
        'Proyecto'
    ].nunique()}"
)


print("\nMAESTRA DE PROYECTOS PARA RELACIÓN CDP")
print("-" * 140)

print(
    f"Filas de la maestra: "
    f"{len(df_maestra_proyectos_cdp)}"
)

print(
    f"Llave AnioFuente-Proyecto única: "
    f"{not df_maestra_proyectos_cdp.duplicated(
        subset=['AnioFuente', 'Proyecto']
    ).any()}"
)


# ------------------------------------------------------------
# MOSTRAR RESPONSABLES MÚLTIPLES
# ------------------------------------------------------------

print("\nRESPONSABLES AGRUPADOS")

responsables_multiples = (
    df_responsables_historico[
        df_responsables_historico[
            "Responsable"
        ].str.contains(
            "/",
            regex=False,
            na=False
        )
    ]
)

print(
    responsables_multiples.to_string(
        index=False,
        max_colwidth=100
    )
)


# ------------------------------------------------------------
# COBERTURA POR AÑO
# ------------------------------------------------------------

print("\nCOBERTURA DE LA MAESTRA CDP")

resumen_cobertura = (
    df_maestra_proyectos_cdp
    .groupby("AnioFuente")
    .agg(
        CantidadProyectos=(
            "Proyecto",
            "count"
        ),
        ConSubsecretaria=(
            "TieneSubsecretaria",
            "sum"
        ),
        ConResponsable=(
            "TieneResponsable",
            "sum"
        )
    )
    .reset_index()
)

print(
    resumen_cobertura.to_string(
        index=False
    )
)


print("\n" + "=" * 140)

if not subsecretarias_unicas:

    print(
        "ADVERTENCIA: la tabla de subsecretarías "
        "no quedó con una fila por proyecto."
    )

elif not responsables_unicos:

    print(
        "ADVERTENCIA: la tabla de responsables "
        "no quedó con una fila por proyecto."
    )

else:

    print(
        "ESTADO: responsables múltiples agrupados con '/'. "
        "Se construyeron la maestra CDP y el histórico "
        "sin duplicar registros por responsable."
    )
```

```
# ============================================================
# PROCESOS POR ESTADO - PASO 3
# LIMPIAR, ESTANDARIZAR Y CONSOLIDAR
# ============================================================

import pandas as pd
import numpy as np
import re


print("=" * 145)
print("LIMPIEZA Y CONSOLIDACIÓN - PROCESOS POR ESTADO")
print("=" * 145)


# ------------------------------------------------------------
# 1. MAPA DE NOMBRES ESTÁNDAR
# ------------------------------------------------------------

mapa_columnas_procesos = {
    "Sec Ejecutora": "SeccionEjecutora",
    "Descripcion": "DescripcionSeccionEjecutora",
    "Secretaria Necesidad": "SecretariaNecesidad",
    "Descripcion.1": "DescripcionSecretariaNecesidad",
    "Necesidad": "Necesidad",
    "Estudio Previo": "EstudioPrevio",
    "SolPed": "SolicitudPedido",
    "Contrato": "Contrato",
    "Objeto": "Objeto",
    "Etapa": "CodigoEtapa",
    "Etapa Actual": "EtapaActual",
    "Estado": "CodigoEstado",
    "Desc Estado": "DescripcionEstado",
    "Actividad": "CodigoActividad",
    "Descripción Act": "DescripcionActividad",
    "Usuario": "Usuario",
    "Nombre": "NombreResponsable",
    "FReal de Inicio": "FechaRealInicio",
    "Duracion": "Duracion",
    "FTerminacion Planificada": "FechaTerminacionPlanificada",
    "Modalidad": "Modalidad",
    "Causal": "Causal"
}


# ------------------------------------------------------------
# 2. FUNCIONES DE LIMPIEZA
# ------------------------------------------------------------

def limpiar_vacio(valor):

    if valor is None:
        return pd.NA

    try:
        if pd.isna(valor):
            return pd.NA
    except (TypeError, ValueError):
        pass

    texto = str(valor).strip()

    if texto == "":
        return pd.NA

    return texto


def limpiar_codigo_puro(valor):
    """
    Conserva únicamente valores completamente numéricos.
    Elimina .0 y devuelve texto.
    No extrae números contenidos dentro de textos.
    """

    if valor is None:
        return pd.NA

    try:
        if pd.isna(valor):
            return pd.NA
    except (TypeError, ValueError):
        pass

    if isinstance(valor, (int, np.integer)):
        return str(int(valor))

    if isinstance(valor, (float, np.floating)):

        if float(valor).is_integer():
            return str(int(valor))

        return str(valor)

    texto = str(valor).strip()

    if re.fullmatch(r"\d+\.0", texto):
        return texto[:-2]

    if re.fullmatch(r"\d+", texto):
        return texto

    return pd.NA


def limpiar_texto(valor):

    valor = limpiar_vacio(valor)

    if pd.isna(valor):
        return pd.NA

    texto = str(valor)

    texto = texto.replace("\r\n", "\n")
    texto = texto.replace("\r", "\n")

    lineas = [
        linea.strip()
        for linea in texto.split("\n")
    ]

    texto = "\n".join(lineas).strip()

    return texto if texto else pd.NA


# ------------------------------------------------------------
# 3. CONFIGURAR TIPOS DE COLUMNAS
# ------------------------------------------------------------

columnas_codigo = [
    "SeccionEjecutora",
    "SecretariaNecesidad",
    "Necesidad",
    "EstudioPrevio",
    "SolicitudPedido",
    "Contrato",
    "CodigoEstado",
    "CodigoActividad"
]

columnas_texto = [
    "DescripcionSeccionEjecutora",
    "DescripcionSecretariaNecesidad",
    "Objeto",
    "CodigoEtapa",
    "EtapaActual",
    "DescripcionEstado",
    "DescripcionActividad",
    "Usuario",
    "NombreResponsable",
    "Modalidad",
    "Causal"
]

columnas_fecha = [
    "FechaRealInicio",
    "FechaTerminacionPlanificada"
]


# ------------------------------------------------------------
# 4. LIMPIAR CADA AÑO
# ------------------------------------------------------------

procesos_limpios_por_anio = {}
resumen_limpieza = []

for anio in ["2023", "2024", "2025", "2026"]:

    print("\n" + "=" * 145)
    print(f"AÑO: {anio}")
    print("=" * 145)

    df_origen = procesos_por_anio[anio].copy()

    # Renombrar columnas
    df_limpio = df_origen.rename(
        columns=mapa_columnas_procesos
    )

    # Códigos como texto
    for columna in columnas_codigo:

        df_limpio[columna] = (
            df_limpio[columna]
            .apply(limpiar_codigo_puro)
            .astype("string")
        )

    # Textos
    for columna in columnas_texto:

        df_limpio[columna] = (
            df_limpio[columna]
            .apply(limpiar_texto)
            .astype("string")
        )

    # Fechas
    for columna in columnas_fecha:

        df_limpio[columna] = pd.to_datetime(
            df_limpio[columna],
            errors="coerce"
        )

    # Duración numérica
    df_limpio["Duracion"] = pd.to_numeric(
        df_limpio["Duracion"],
        errors="coerce"
    )

    # Trazabilidad
    df_limpio["AnioFuente"] = (
        pd.to_numeric(
            df_limpio["AnioFuente"],
            errors="coerce"
        )
        .astype("Int64")
    )

    df_limpio["FilaExcelFuente"] = (
        pd.to_numeric(
            df_limpio["FilaExcelFuente"],
            errors="coerce"
        )
        .astype("Int64")
    )

    df_limpio["ArchivoFuente"] = (
        df_limpio["ArchivoFuente"]
        .apply(limpiar_texto)
        .astype("string")
    )

    procesos_limpios_por_anio[anio] = (
        df_limpio.copy()
    )

    resumen_limpieza.append({
        "Anio": anio,
        "RegistrosAntes": len(df_origen),
        "RegistrosDespues": len(df_limpio),
        "ColumnasResultado": len(df_limpio.columns)
    })

    print(f"Registros antes: {len(df_origen)}")
    print(f"Registros después: {len(df_limpio)}")
    print(f"Columnas: {len(df_limpio.columns)}")


# ------------------------------------------------------------
# 5. VALIDAR ESTRUCTURA IDÉNTICA
# ------------------------------------------------------------

columnas_base = list(
    procesos_limpios_por_anio["2023"].columns
)

estructura_identica = all(
    list(procesos_limpios_por_anio[anio].columns)
    == columnas_base
    for anio in ["2023", "2024", "2025", "2026"]
)


# ------------------------------------------------------------
# 6. CONSOLIDAR
# ------------------------------------------------------------

df_procesos_estado_consolidado = pd.concat(
    [
        procesos_limpios_por_anio["2023"],
        procesos_limpios_por_anio["2024"],
        procesos_limpios_por_anio["2025"],
        procesos_limpios_por_anio["2026"]
    ],
    ignore_index=True
)


# ------------------------------------------------------------
# 7. ORDENAR Y CREAR IDENTIFICADOR
# ------------------------------------------------------------

df_procesos_estado_consolidado = (
    df_procesos_estado_consolidado
    .sort_values(
        by=[
            "AnioFuente",
            "FilaExcelFuente"
        ]
    )
    .reset_index(drop=True)
)

df_procesos_estado_consolidado.insert(
    0,
    "IdRegistro",
    range(
        1,
        len(df_procesos_estado_consolidado) + 1
    )
)

df_procesos_estado_consolidado[
    "IdRegistro"
] = (
    df_procesos_estado_consolidado[
        "IdRegistro"
    ]
    .astype("Int64")
)


# ------------------------------------------------------------
# 8. IDENTIFICAR DUPLICADOS EXACTOS DE NEGOCIO
# ------------------------------------------------------------

columnas_negocio = [
    columna
    for columna in df_procesos_estado_consolidado.columns
    if columna not in [
        "IdRegistro",
        "AnioFuente",
        "ArchivoFuente",
        "FilaExcelFuente"
    ]
]

mascara_duplicados = (
    df_procesos_estado_consolidado
    .duplicated(
        subset=columnas_negocio,
        keep=False
    )
)

df_duplicados_procesos_estado = (
    df_procesos_estado_consolidado.loc[
        mascara_duplicados
    ]
    .copy()
)


# ------------------------------------------------------------
# 9. CONTROL TEMPORAL
# ------------------------------------------------------------

mascara_inicio_otro_anio = (
    df_procesos_estado_consolidado[
        "FechaRealInicio"
    ].notna()
    &
    (
        df_procesos_estado_consolidado[
            "FechaRealInicio"
        ].dt.year
        !=
        df_procesos_estado_consolidado[
            "AnioFuente"
        ]
    )
)

df_inicios_otro_anio_procesos = (
    df_procesos_estado_consolidado.loc[
        mascara_inicio_otro_anio,
        [
            "IdRegistro",
            "AnioFuente",
            "FechaRealInicio",
            "Necesidad",
            "EstudioPrevio",
            "Contrato",
            "Objeto",
            "ArchivoFuente",
            "FilaExcelFuente"
        ]
    ]
    .copy()
)


# ------------------------------------------------------------
# 10. RESUMEN POR AÑO
# ------------------------------------------------------------

resumen_consolidado = (
    df_procesos_estado_consolidado
    .groupby("AnioFuente")
    .size()
    .reset_index(name="CantidadRegistros")
)

print("\n" + "=" * 145)
print("RESUMEN DEL CONSOLIDADO")
print("=" * 145)

print(
    resumen_consolidado.to_string(
        index=False
    )
)

print(
    f"\nTotal consolidado: "
    f"{len(df_procesos_estado_consolidado)}"
)

print(
    f"Cantidad de columnas: "
    f"{len(df_procesos_estado_consolidado.columns)}"
)

print(
    f"Duplicados exactos: "
    f"{len(df_duplicados_procesos_estado)}"
)

print(
    f"Fechas reales de inicio diferentes al año fuente: "
    f"{len(df_inicios_otro_anio_procesos)}"
)


# ------------------------------------------------------------
# 11. MOSTRAR MUESTRA
# ------------------------------------------------------------

columnas_muestra = [
    "IdRegistro",
    "AnioFuente",
    "Necesidad",
    "EstudioPrevio",
    "SolicitudPedido",
    "Contrato",
    "CodigoEtapa",
    "EtapaActual",
    "DescripcionEstado",
    "FechaRealInicio",
    "FechaTerminacionPlanificada",
    "FilaExcelFuente"
]

print("\nPRIMEROS 10 REGISTROS")

print(
    df_procesos_estado_consolidado[
        columnas_muestra
    ]
    .head(10)
    .to_string(
        index=False,
        max_colwidth=70
    )
)


# ------------------------------------------------------------
# 12. VALIDACIÓN FINAL
# ------------------------------------------------------------

conteos_esperados = {
    2023: 27,
    2024: 61,
    2025: 91,
    2026: 60
}

conteos_correctos = all(
    int(
        resumen_consolidado.loc[
            resumen_consolidado["AnioFuente"] == anio,
            "CantidadRegistros"
        ].iloc[0]
    ) == cantidad
    for anio, cantidad in conteos_esperados.items()
)

print("\n" + "=" * 145)

if not estructura_identica:

    print(
        "ADVERTENCIA: los cuatro años no tienen "
        "la misma estructura."
    )

elif len(df_procesos_estado_consolidado) != 239:

    print(
        "ADVERTENCIA: el consolidado no contiene "
        "los 239 registros esperados."
    )

elif not conteos_correctos:

    print(
        "ADVERTENCIA: algún año no conserva "
        "su cantidad esperada."
    )

elif not df_duplicados_procesos_estado.empty:

    print(
        "ADVERTENCIA: existen duplicados exactos "
        "que deben revisarse."
    )

else:

    print(
        "ESTADO: los cuatro años fueron limpiados y "
        "consolidados correctamente en 239 registros. "
        "No se encontraron duplicados exactos y todavía "
        "no se exportaron archivos."
    )
```

```
# ============================================================
# AJUSTES Y RESERVAS - PASO 2
# COMPARAR COLUMNAS, HOMOLOGAR NOMBRES Y PERFILAR COMPLETITUD
# ============================================================

import pandas as pd
import numpy as np
import re
import unicodedata


print("=" * 145)
print("COMPARACIÓN DE ESTRUCTURA Y COMPLETITUD - AJUSTES Y RESERVAS")
print("=" * 145)


# ------------------------------------------------------------
# 1. NORMALIZAR ENCABEZADOS SOLO PARA COMPARACIÓN
# ------------------------------------------------------------

def normalizar_encabezado_ajustes(nombre):

    nombre = str(nombre).strip()
    nombre = nombre.replace("\n", " ")

    nombre = unicodedata.normalize(
        "NFKD",
        nombre
    )

    nombre = "".join(
        caracter
        for caracter in nombre
        if not unicodedata.combining(caracter)
    )

    nombre = nombre.lower()

    nombre = re.sub(
        r"[^a-z0-9]+",
        " ",
        nombre
    )

    nombre = re.sub(
        r"\s+",
        " ",
        nombre
    ).strip()

    return nombre


# ------------------------------------------------------------
# 2. LEER LOS TRES ARCHIVOS
# ------------------------------------------------------------

ajustes_por_anio = {}
resumen_columnas = []
perfil_completitud_ajustes = []
errores_ajustes_paso2 = []

for _, fila in df_ajustes_archivos.iterrows():

    anio = str(fila["Anio"])
    archivo = fila["Archivo"]
    ruta = fila["RutaCompleta"]

    print("\n" + "=" * 145)
    print(f"AÑO: {anio}")
    print(f"ARCHIVO: {archivo}")
    print("=" * 145)

    try:

        df_anio = pd.read_excel(
            ruta,
            sheet_name="741",
            header=0,
            engine="openpyxl"
        )

        # Eliminar solo filas completamente vacías
        df_anio = (
            df_anio
            .dropna(axis=0, how="all")
            .reset_index(drop=True)
        )

        # Mantener todas las columnas originales,
        # incluso las completamente vacías
        ajustes_por_anio[anio] = df_anio.copy()

        print(f"Registros reales: {len(df_anio)}")
        print(f"Columnas originales: {len(df_anio.columns)}")

        print("\nCOLUMNAS ORIGINALES")

        for posicion, columna in enumerate(
            df_anio.columns,
            start=1
        ):

            serie = df_anio[columna].replace(
                r"^\s*$",
                np.nan,
                regex=True
            )

            con_dato = int(
                serie.notna().sum()
            )

            vacios = int(
                serie.isna().sum()
            )

            porcentaje = (
                con_dato / len(df_anio) * 100
                if len(df_anio) > 0
                else 0
            )

            nombre_normalizado = (
                normalizar_encabezado_ajustes(
                    columna
                )
            )

            resumen_columnas.append({
                "Anio": anio,
                "Posicion": posicion,
                "ColumnaOriginal": str(columna),
                "ColumnaNormalizada": nombre_normalizado,
                "ConDato": con_dato,
                "Vacios": vacios,
                "PorcentajeConDato": round(
                    porcentaje,
                    2
                )
            })

            perfil_completitud_ajustes.append({
                "Anio": anio,
                "Archivo": archivo,
                "ColumnaOriginal": str(columna),
                "ColumnaNormalizada": nombre_normalizado,
                "TotalFilas": len(df_anio),
                "ConDato": con_dato,
                "Vacios": vacios,
                "PorcentajeConDato": round(
                    porcentaje,
                    2
                ),
                "TipoPandas": str(
                    df_anio[columna].dtype
                )
            })

            print(
                f"{posicion:02d}. "
                f"{str(columna):60} | "
                f"Con dato: {con_dato:4d} | "
                f"Vacíos: {vacios:4d} | "
                f"Completitud: {porcentaje:6.2f}%"
            )

    except Exception as error:

        print(f"ERROR: {error}")

        errores_ajustes_paso2.append({
            "Anio": anio,
            "Archivo": archivo,
            "Error": str(error)
        })


# ------------------------------------------------------------
# 3. CREAR DATAFRAMES DE PERFIL
# ------------------------------------------------------------

df_columnas_ajustes = pd.DataFrame(
    resumen_columnas
)

df_perfil_ajustes_reservas = pd.DataFrame(
    perfil_completitud_ajustes
)


# ------------------------------------------------------------
# 4. MATRIZ COMPARATIVA DE COLUMNAS NORMALIZADAS
# ------------------------------------------------------------

print("\n" + "=" * 145)
print("MATRIZ DE PRESENCIA DE COLUMNAS POR AÑO")
print("=" * 145)

columnas_normalizadas = sorted(
    df_columnas_ajustes[
        "ColumnaNormalizada"
    ].unique()
)

matriz_presencia = []

for columna_normalizada in columnas_normalizadas:

    registro = {
        "ColumnaNormalizada": columna_normalizada
    }

    for anio in ["2023", "2024", "2025"]:

        coincidencias = (
            df_columnas_ajustes[
                (
                    df_columnas_ajustes["Anio"] == anio
                )
                &
                (
                    df_columnas_ajustes[
                        "ColumnaNormalizada"
                    ] == columna_normalizada
                )
            ]["ColumnaOriginal"]
            .tolist()
        )

        registro[f"Columna{anio}"] = (
            " | ".join(coincidencias)
            if coincidencias
            else None
        )

    matriz_presencia.append(registro)


df_matriz_columnas_ajustes = pd.DataFrame(
    matriz_presencia
)

print(
    df_matriz_columnas_ajustes.to_string(
        index=False,
        max_colwidth=80
    )
)


# ------------------------------------------------------------
# 5. COLUMNAS QUE NO ESTÁN EN LOS TRES AÑOS
# ------------------------------------------------------------

print("\n" + "=" * 145)
print("COLUMNAS CON DIFERENCIAS ENTRE AÑOS")
print("=" * 145)

columnas_diferentes = (
    df_matriz_columnas_ajustes[
        df_matriz_columnas_ajustes[
            [
                "Columna2023",
                "Columna2024",
                "Columna2025"
            ]
        ]
        .isna()
        .any(axis=1)
    ]
)

if columnas_diferentes.empty:

    print(
        "Todas las columnas están presentes "
        "en los tres años."
    )

else:

    print(
        columnas_diferentes.to_string(
            index=False,
            max_colwidth=100
        )
    )


# ------------------------------------------------------------
# 6. COLUMNAS COMPLETAMENTE VACÍAS
# ------------------------------------------------------------

print("\n" + "=" * 145)
print("COLUMNAS COMPLETAMENTE VACÍAS")
print("=" * 145)

columnas_vacias = (
    df_perfil_ajustes_reservas[
        df_perfil_ajustes_reservas[
            "ConDato"
        ] == 0
    ]
)

if columnas_vacias.empty:

    print(
        "No se encontraron columnas completamente vacías."
    )

else:

    print(
        columnas_vacias[
            [
                "Anio",
                "ColumnaOriginal",
                "TotalFilas"
            ]
        ].to_string(
            index=False
        )
    )


# ------------------------------------------------------------
# 7. VALIDAR DUPLICADOS EXACTOS
# ------------------------------------------------------------

print("\n" + "=" * 145)
print("DUPLICADOS EXACTOS POR AÑO")
print("=" * 145)

for anio, df_anio in ajustes_por_anio.items():

    duplicados = int(
        df_anio.duplicated(
            keep=False
        ).sum()
    )

    print(
        f"Año {anio}: "
        f"{duplicados} filas involucradas "
        f"en duplicados exactos"
    )


# ------------------------------------------------------------
# 8. RESUMEN
# ------------------------------------------------------------

print("\n" + "=" * 145)
print("RESUMEN DE REGISTROS")
print("=" * 145)

total_registros = 0

for anio, df_anio in ajustes_por_anio.items():

    cantidad = len(df_anio)
    total_registros += cantidad

    print(
        f"Año {anio}: {cantidad} registros"
    )

print(
    f"\nTotal preliminar: "
    f"{total_registros} registros"
)


# ------------------------------------------------------------
# 9. VALIDACIÓN FINAL
# ------------------------------------------------------------

print("\n" + "=" * 145)

if errores_ajustes_paso2:

    print(
        "ESTADO: se presentaron errores durante "
        "la comparación."
    )

    for error in errores_ajustes_paso2:

        print(
            f"- Año {error['Anio']}: "
            f"{error['Error']}"
        )

elif len(ajustes_por_anio) != 3:

    print(
        "ADVERTENCIA: no se procesaron "
        "los tres años."
    )

else:

    print(
        "ESTADO: los tres años fueron comparados y "
        "perfilados correctamente. Todavía no se "
        "seleccionaron, limpiaron ni unieron columnas."
    )
```

```
# ============================================================
# AJUSTES Y RESERVAS - PASO 4
# HOMOLOGAR, LIMPIAR Y CONSOLIDAR LOS TRES AÑOS
# ============================================================

import pandas as pd
import numpy as np
import re
import unicodedata


print("=" * 150)
print("HOMOLOGACIÓN Y CONSOLIDACIÓN - AJUSTES Y RESERVAS")
print("=" * 150)


# ------------------------------------------------------------
# 1. FUNCIONES AUXILIARES
# ------------------------------------------------------------

def normalizar_columna(nombre):

    texto = str(nombre).strip()

    texto = unicodedata.normalize(
        "NFKD",
        texto
    )

    texto = "".join(
        caracter
        for caracter in texto
        if not unicodedata.combining(caracter)
    )

    texto = texto.lower()
    texto = re.sub(r"[^a-z0-9]+", "", texto)

    return texto


def localizar_columna(df, candidatos):

    if isinstance(candidatos, str):
        candidatos = [candidatos]

    columnas_normalizadas = {
        normalizar_columna(columna): columna
        for columna in df.columns
    }

    for candidato in candidatos:

        candidato_normalizado = normalizar_columna(
            candidato
        )

        if candidato_normalizado in columnas_normalizadas:

            return columnas_normalizadas[
                candidato_normalizado
            ]

    return None


def obtener_serie(df, candidatos):

    columna = localizar_columna(
        df,
        candidatos
    )

    if columna is None:

        return pd.Series(
            pd.NA,
            index=df.index,
            dtype="object"
        )

    return df[columna].copy()


def es_vacio(valor):

    if valor is None:
        return True

    try:
        if pd.isna(valor):
            return True
    except (TypeError, ValueError):
        pass

    return str(valor).strip() == ""


def limpiar_codigo(valor):
    """
    Conserva el contenido como texto y elimina únicamente
    la terminación .0 cuando representa un entero.
    """

    if es_vacio(valor):
        return pd.NA

    if isinstance(valor, (int, np.integer)):
        return str(int(valor))

    if isinstance(valor, (float, np.floating)):

        if float(valor).is_integer():
            return str(int(valor))

        return str(valor)

    texto = str(valor).strip()

    if re.fullmatch(r"-?\d+\.0", texto):
        return texto[:-2]

    return texto


def limpiar_texto(valor):

    if es_vacio(valor):
        return pd.NA

    texto = str(valor).strip()

    texto = texto.replace("\r\n", "\n")
    texto = texto.replace("\r", "\n")

    lineas = [
        linea.strip()
        for linea in texto.split("\n")
    ]

    texto = "\n".join(lineas).strip()

    return texto if texto else pd.NA


def limpiar_numero(valor):

    if es_vacio(valor):
        return np.nan

    if isinstance(valor, (int, float, np.integer, np.floating)):
        return float(valor)

    texto = str(valor).strip()

    texto = texto.replace(" ", "")
    texto = texto.replace("$", "")

    # Formato colombiano: 1.234.567,89
    if "," in texto:

        texto = texto.replace(".", "")
        texto = texto.replace(",", ".")

    try:
        return float(texto)

    except (ValueError, TypeError):
        return np.nan


# ------------------------------------------------------------
# 2. MAPA DE CAMPOS ESTÁNDAR
# ------------------------------------------------------------

mapa_ajustes = {
    "Consecutivo": ["Cons"],
    "NumeroDocumentoReferencia": [
        "Nº docum.refer.",
        "Nº docum. refer."
    ],
    "Proveedor": ["Proveedor"],
    "NombreProveedor": ["Nombre 1"],
    "IdentificacionFiscal": ["Nº ident.fis.1"],
    "PosicionReferencia": [
        "Pos.referencia",
        "Pos. referencia"
    ],
    "NumeroReparto": ["Nº de reparto"],
    "Ledger": ["Ledger"],
    "NumeroImputacion": ["Número de imputación"],
    "NumeroContrato": ["Contrato marco"],
    "Ejercicio": ["Ejercicio"],
    "FechaContabilizacion": [
        "Fe.contabilización",
        "Fe. contabilización",
        "Fecha contabilización"
    ],
    "TipoValor": ["Tipo de valor"],
    "SolicitudPedido": ["No. Solicitud Pedido"],
    "PosicionSolicitudPedido": [
        "Pos. Solicitud Pedido"
    ],
    "NumeroCDP": ["No.  CDP"],
    "PosicionCDP": ["Pos. CDP"],
    "NivelArrastreSaldo": ["Niv.arr.saldo"],
    "Fondo": ["Fondo"],
    "DescripcionFondo": ["Descripción"],
    "CentroGestor": ["Centro gestor"],
    "DescripcionCentroGestor": ["Descripción.1"],
    "PosicionPresupuestaria": [
        "Posición presupuestaria"
    ],
    "DescripcionPosicionPresupuestaria": [
        "Descripción.2"
    ],
    "AreaFuncional": ["Área funcional"],
    "Proyecto": ["Programa de financiación"],
    "DescripcionProyecto": ["Descripción.3"],
    "TotalAsignado": ["Total asignado"],
    "DescripcionMovimiento": ["Ver Descripcion"],
    "Observacion1": [
        "Observación 1",
        "Observación 1 (Seleccionar Opción)"
    ],
    "Observacion2Causal": [
        "Observación 2 (Causal de la Resolución 091)",
        "Observación 2 (Causal de la Resolución 092)"
    ],
    "JustificacionSecretaria": [
        "Justificación de la Secretaría",
        "Justificación de la Secretaría (Seleccionar Opción)"
    ],
    "PorQueNoSeRequiere": [
        "Por que no se requiere"
    ],
    "ComentariosGenerales": [
        "Comentarios Generales"
    ]
}


columnas_codigo_ajustes = [
    "Consecutivo",
    "NumeroDocumentoReferencia",
    "Proveedor",
    "IdentificacionFiscal",
    "PosicionReferencia",
    "NumeroReparto",
    "NumeroImputacion",
    "NumeroContrato",
    "Ejercicio",
    "TipoValor",
    "SolicitudPedido",
    "PosicionSolicitudPedido",
    "NumeroCDP",
    "PosicionCDP",
    "Fondo",
    "CentroGestor",
    "PosicionPresupuestaria",
    "AreaFuncional",
    "Proyecto"
]


columnas_texto_ajustes = [
    "NombreProveedor",
    "Ledger",
    "NivelArrastreSaldo",
    "DescripcionFondo",
    "DescripcionCentroGestor",
    "DescripcionPosicionPresupuestaria",
    "DescripcionProyecto",
    "DescripcionMovimiento",
    "Observacion1",
    "Observacion2Causal",
    "JustificacionSecretaria",
    "PorQueNoSeRequiere",
    "ComentariosGenerales"
]


# ------------------------------------------------------------
# 3. HOMOLOGAR Y LIMPIAR POR AÑO
# ------------------------------------------------------------

ajustes_limpios_por_anio = {}
resumen_limpieza_ajustes = []


for anio in ["2023", "2024", "2025"]:

    print("\n" + "=" * 150)
    print(f"AÑO: {anio}")
    print("=" * 150)

    df_origen = (
        filas_reales_ajustes_por_anio[anio]
        .copy()
    )

    df_resultado = pd.DataFrame(
        index=df_origen.index
    )

    for nombre_estandar, candidatos in mapa_ajustes.items():

        df_resultado[nombre_estandar] = obtener_serie(
            df_origen,
            candidatos
        )

    # Códigos
    for columna in columnas_codigo_ajustes:

        df_resultado[columna] = (
            df_resultado[columna]
            .apply(limpiar_codigo)
            .astype("string")
        )

    # Textos
    for columna in columnas_texto_ajustes:

        df_resultado[columna] = (
            df_resultado[columna]
            .apply(limpiar_texto)
            .astype("string")
        )

    # Fecha
    df_resultado["FechaContabilizacion"] = (
        pd.to_datetime(
            df_resultado["FechaContabilizacion"],
            errors="coerce",
            dayfirst=True
        )
    )

    # Total numérico
    df_resultado["TotalAsignado"] = (
        df_resultado["TotalAsignado"]
        .apply(limpiar_numero)
    )

    # Trazabilidad
    archivo_fuente = (
        config_ajustes_reservas[anio]["Archivo"]
    )

    df_resultado["AnioFuente"] = int(anio)
    df_resultado["ArchivoFuente"] = archivo_fuente

    df_resultado["FilaExcelFuente"] = (
        pd.to_numeric(
            df_origen["_FilaExcel"],
            errors="coerce"
        )
        .astype("Int64")
    )

    df_resultado["AnioFuente"] = (
        df_resultado["AnioFuente"]
        .astype("Int64")
    )

    ajustes_limpios_por_anio[anio] = (
        df_resultado.reset_index(drop=True)
    )

    resumen_limpieza_ajustes.append({
        "Anio": anio,
        "RegistrosEntrada": len(df_origen),
        "RegistrosSalida": len(df_resultado),
        "ColumnasSalida": len(df_resultado.columns),
        "FechasInvalidas": int(
            df_resultado[
                "FechaContabilizacion"
            ].isna().sum()
        )
    })

    print(
        f"Registros de entrada: {len(df_origen)}"
    )

    print(
        f"Registros de salida: {len(df_resultado)}"
    )

    print(
        f"Columnas de salida: {len(df_resultado.columns)}"
    )

    print(
        "Fechas inválidas o vacías: "
        f"{df_resultado['FechaContabilizacion'].isna().sum()}"
    )


# ------------------------------------------------------------
# 4. VALIDAR ESTRUCTURA IDÉNTICA
# ------------------------------------------------------------

columnas_base = list(
    ajustes_limpios_por_anio["2023"].columns
)

estructura_identica = all(
    list(ajustes_limpios_por_anio[anio].columns)
    == columnas_base
    for anio in ["2023", "2024", "2025"]
)


# ------------------------------------------------------------
# 5. CONSOLIDAR
# ------------------------------------------------------------

df_ajustes_reservas_consolidado = pd.concat(
    [
        ajustes_limpios_por_anio["2023"],
        ajustes_limpios_por_anio["2024"],
        ajustes_limpios_por_anio["2025"]
    ],
    ignore_index=True
)


df_ajustes_reservas_consolidado = (
    df_ajustes_reservas_consolidado
    .sort_values(
        by=[
            "AnioFuente",
            "FilaExcelFuente"
        ]
    )
    .reset_index(drop=True)
)


# ------------------------------------------------------------
# 6. CREAR IDENTIFICADOR
# ------------------------------------------------------------

df_ajustes_reservas_consolidado.insert(
    0,
    "IdRegistro",
    range(
        1,
        len(df_ajustes_reservas_consolidado) + 1
    )
)

df_ajustes_reservas_consolidado[
    "IdRegistro"
] = (
    df_ajustes_reservas_consolidado[
        "IdRegistro"
    ]
    .astype("Int64")
)


# ------------------------------------------------------------
# 7. VALIDAR DUPLICADOS EXACTOS DE NEGOCIO
# ------------------------------------------------------------

columnas_negocio = [
    columna
    for columna in df_ajustes_reservas_consolidado.columns
    if columna not in [
        "IdRegistro",
        "AnioFuente",
        "ArchivoFuente",
        "FilaExcelFuente"
    ]
]

mascara_duplicados = (
    df_ajustes_reservas_consolidado
    .duplicated(
        subset=columnas_negocio,
        keep=False
    )
)

df_duplicados_ajustes_reservas = (
    df_ajustes_reservas_consolidado.loc[
        mascara_duplicados
    ]
    .copy()
)


# ------------------------------------------------------------
# 8. VALIDACIÓN TEMPORAL
# ------------------------------------------------------------

mascara_fecha_otro_anio = (
    df_ajustes_reservas_consolidado[
        "FechaContabilizacion"
    ].notna()
    &
    (
        df_ajustes_reservas_consolidado[
            "FechaContabilizacion"
        ].dt.year
        !=
        df_ajustes_reservas_consolidado[
            "AnioFuente"
        ]
    )
)

df_fechas_otro_anio_ajustes = (
    df_ajustes_reservas_consolidado.loc[
        mascara_fecha_otro_anio
    ]
    .copy()
)


# ------------------------------------------------------------
# 9. RESUMEN
# ------------------------------------------------------------

df_resumen_limpieza_ajustes = pd.DataFrame(
    resumen_limpieza_ajustes
)

print("\n" + "=" * 150)
print("RESUMEN DE LIMPIEZA")
print("=" * 150)

print(
    df_resumen_limpieza_ajustes.to_string(
        index=False
    )
)


resumen_por_anio = (
    df_ajustes_reservas_consolidado
    .groupby("AnioFuente")
    .size()
    .reset_index(name="CantidadRegistros")
)

print("\nREGISTROS CONSOLIDADOS POR AÑO")

print(
    resumen_por_anio.to_string(
        index=False
    )
)

print(
    f"\nTotal consolidado: "
    f"{len(df_ajustes_reservas_consolidado)}"
)

print(
    f"Cantidad de columnas: "
    f"{len(df_ajustes_reservas_consolidado.columns)}"
)

print(
    f"Filas involucradas en duplicados exactos: "
    f"{len(df_duplicados_ajustes_reservas)}"
)

print(
    f"Fechas diferentes al año fuente: "
    f"{len(df_fechas_otro_anio_ajustes)}"
)


# ------------------------------------------------------------
# 10. MUESTRA
# ------------------------------------------------------------

columnas_muestra = [
    "IdRegistro",
    "AnioFuente",
    "NumeroDocumentoReferencia",
    "Proveedor",
    "NumeroContrato",
    "SolicitudPedido",
    "NumeroCDP",
    "Fondo",
    "Proyecto",
    "TotalAsignado",
    "Observacion1",
    "FilaExcelFuente"
]

print("\nPRIMEROS 10 REGISTROS")

print(
    df_ajustes_reservas_consolidado[
        columnas_muestra
    ]
    .head(10)
    .to_string(
        index=False,
        max_colwidth=70
    )
)


# ------------------------------------------------------------
# 11. VALIDACIÓN FINAL
# ------------------------------------------------------------

conteos_esperados = {
    2023: 282,
    2024: 340,
    2025: 389
}

conteos_correctos = all(
    int(
        resumen_por_anio.loc[
            resumen_por_anio["AnioFuente"] == anio,
            "CantidadRegistros"
        ].iloc[0]
    ) == cantidad
    for anio, cantidad in conteos_esperados.items()
)


print("\n" + "=" * 150)

if not estructura_identica:

    print(
        "ADVERTENCIA: los tres años no tienen "
        "la misma estructura."
    )

elif len(df_ajustes_reservas_consolidado) != 1011:

    print(
        "ADVERTENCIA: el consolidado no contiene "
        "los 1.011 registros esperados."
    )

elif not conteos_correctos:

    print(
        "ADVERTENCIA: algún año no conserva "
        "la cantidad esperada."
    )

else:

    print(
        "ESTADO: los tres años fueron homologados, limpiados "
        "y consolidados correctamente en 1.011 registros. "
        "Las cuatro filas residuales fueron excluidas y "
        "todavía no se exportaron archivos."
    )
```

```
# ============================================================
# AJUSTES Y RESERVAS - PASO 5
# ANALIZAR FECHA DE CONTABILIZACIÓN VS AÑO FUENTE
# ============================================================

import pandas as pd


print("=" * 140)
print("ANÁLISIS TEMPORAL - AJUSTES Y RESERVAS")
print("=" * 140)


# ------------------------------------------------------------
# 1. CREAR AÑO DE CONTABILIZACIÓN
# ------------------------------------------------------------

df_control_fechas_ajustes = (
    df_ajustes_reservas_consolidado.copy()
)

df_control_fechas_ajustes[
    "AnioContabilizacion"
] = (
    df_control_fechas_ajustes[
        "FechaContabilizacion"
    ]
    .dt.year
    .astype("Int64")
)


# ------------------------------------------------------------
# 2. RESUMEN AÑO FUENTE VS AÑO CONTABILIZACIÓN
# ------------------------------------------------------------

resumen_temporal_ajustes = (
    df_control_fechas_ajustes
    .groupby(
        [
            "AnioFuente",
            "AnioContabilizacion"
        ],
        dropna=False
    )
    .size()
    .reset_index(
        name="CantidadRegistros"
    )
    .sort_values(
        [
            "AnioFuente",
            "AnioContabilizacion"
        ]
    )
)


print("\nDISTRIBUCIÓN GENERAL\n")

print(
    resumen_temporal_ajustes.to_string(
        index=False
    )
)


# ------------------------------------------------------------
# 3. CLASIFICAR RELACIÓN TEMPORAL
# ------------------------------------------------------------

df_control_fechas_ajustes[
    "RelacionTemporal"
] = "Mismo año"

df_control_fechas_ajustes.loc[
    df_control_fechas_ajustes[
        "AnioContabilizacion"
    ]
    <
    df_control_fechas_ajustes[
        "AnioFuente"
    ],
    "RelacionTemporal"
] = "Año anterior"

df_control_fechas_ajustes.loc[
    df_control_fechas_ajustes[
        "AnioContabilizacion"
    ]
    >
    df_control_fechas_ajustes[
        "AnioFuente"
    ],
    "RelacionTemporal"
] = "Año posterior"


resumen_relacion_temporal = (
    df_control_fechas_ajustes
    .groupby(
        [
            "AnioFuente",
            "RelacionTemporal"
        ]
    )
    .size()
    .reset_index(
        name="CantidadRegistros"
    )
)


print("\n" + "=" * 140)
print("CLASIFICACIÓN TEMPORAL")
print("=" * 140)

print(
    resumen_relacion_temporal.to_string(
        index=False
    )
)


# ------------------------------------------------------------
# 4. IDENTIFICAR FECHAS DE AÑOS POSTERIORES
# Estas sí requerirían mayor atención.
# ------------------------------------------------------------

df_fechas_posteriores_ajustes = (
    df_control_fechas_ajustes[
        df_control_fechas_ajustes[
            "AnioContabilizacion"
        ]
        >
        df_control_fechas_ajustes[
            "AnioFuente"
        ]
    ]
    [
        [
            "IdRegistro",
            "AnioFuente",
            "FechaContabilizacion",
            "AnioContabilizacion",
            "NumeroDocumentoReferencia",
            "NumeroContrato",
            "NumeroCDP",
            "Proyecto",
            "TotalAsignado",
            "ArchivoFuente",
            "FilaExcelFuente"
        ]
    ]
    .copy()
)


print("\n" + "=" * 140)
print("FECHAS DE CONTABILIZACIÓN POSTERIORES AL AÑO FUENTE")
print("=" * 140)

if df_fechas_posteriores_ajustes.empty:

    print(
        "No se encontraron fechas de contabilización "
        "posteriores al año del archivo."
    )

else:

    print(
        df_fechas_posteriores_ajustes.to_string(
            index=False,
            max_colwidth=80
        )
    )


# ------------------------------------------------------------
# 5. EJEMPLOS DE FECHAS DEL AÑO ANTERIOR
# ------------------------------------------------------------

df_fechas_anteriores_ajustes = (
    df_control_fechas_ajustes[
        df_control_fechas_ajustes[
            "AnioContabilizacion"
        ]
        <
        df_control_fechas_ajustes[
            "AnioFuente"
        ]
    ]
    .copy()
)


print("\n" + "=" * 140)
print("MUESTRA DE CONTABILIZACIONES DE AÑOS ANTERIORES")
print("=" * 140)

print(
    df_fechas_anteriores_ajustes[
        [
            "IdRegistro",
            "AnioFuente",
            "FechaContabilizacion",
            "NumeroContrato",
            "NumeroCDP",
            "Fondo",
            "DescripcionFondo",
            "Observacion1",
            "FilaExcelFuente"
        ]
    ]
    .head(20)
    .to_string(
        index=False,
        max_colwidth=80
    )
)


# ------------------------------------------------------------
# 6. VALIDACIÓN FINAL
# ------------------------------------------------------------

cantidad_diferente = int(
    (
        df_control_fechas_ajustes[
            "AnioContabilizacion"
        ]
        !=
        df_control_fechas_ajustes[
            "AnioFuente"
        ]
    ).sum()
)

cantidad_posterior = len(
    df_fechas_posteriores_ajustes
)


print("\n" + "=" * 140)

print(
    f"Fechas diferentes al año fuente: "
    f"{cantidad_diferente}"
)

print(
    f"Fechas de años posteriores: "
    f"{cantidad_posterior}"
)

print(
    f"Fechas de años anteriores: "
    f"{len(df_fechas_anteriores_ajustes)}"
)

print("\nESTADO: análisis temporal terminado. "
      "No se modificaron los 1.011 registros.")
```
