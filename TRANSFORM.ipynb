```
# ============================================================
# PASO 4 - INVENTARIAR EL CONTENIDO DEL ZIP SIN DESCOMPRIMIR
# ============================================================

import zipfile
import os
from collections import Counter

ruta_zip = "/content/OneDrive_2026-06-05.zip"

print("=" * 90)
print("INVENTARIO DEL CONTENIDO DEL ZIP")
print("=" * 90)

with zipfile.ZipFile(ruta_zip, "r") as zip_ref:

    contenido = zip_ref.infolist()

    archivos = [
        item for item in contenido
        if not item.is_dir()
    ]

    carpetas = [
        item for item in contenido
        if item.is_dir()
    ]

    print(f"Cantidad de carpetas: {len(carpetas)}")
    print(f"Cantidad de archivos: {len(archivos)}")

    extensiones = Counter()

    for item in archivos:
        extension = os.path.splitext(item.filename)[1].lower()

        if extension == "":
            extension = "[sin extensión]"

        extensiones[extension] += 1

    print("\nARCHIVOS POR EXTENSIÓN")

    for extension, cantidad in extensiones.most_common():
        print(f"{extension}: {cantidad}")

    print("\nLISTADO DE ARCHIVOS")

    for numero, item in enumerate(archivos, start=1):
        tamaño_mb = item.file_size / (1024 * 1024)

        print(
            f"{numero:03d}. "
            f"{item.filename} | "
            f"{tamaño_mb:.2f} MB"
        )
```

```
# ============================================================
# PASO 5 - EXTRAER LOS ARCHIVOS ORIGINALES DEL ZIP
# ============================================================

import os
import zipfile
import shutil

# ------------------------------------------------------------
# RUTAS
# ------------------------------------------------------------

ruta_zip = "/content/OneDrive_2026-06-05.zip"
carpeta_extraccion = "/content/01_ORIGINALES"

print("=" * 90)
print("EXTRACCIÓN CONTROLADA DE LOS ARCHIVOS")
print("=" * 90)

# ------------------------------------------------------------
# LIMPIAR LA CARPETA SI YA EXISTÍA DE UNA EJECUCIÓN ANTERIOR
# ------------------------------------------------------------

if os.path.exists(carpeta_extraccion):
    shutil.rmtree(carpeta_extraccion)

os.makedirs(carpeta_extraccion, exist_ok=True)

# ------------------------------------------------------------
# EXTRAER EL ZIP
# ------------------------------------------------------------

with zipfile.ZipFile(ruta_zip, "r") as zip_ref:
    zip_ref.extractall(carpeta_extraccion)

print(f"Carpeta de extracción: {carpeta_extraccion}")

# ------------------------------------------------------------
# VERIFICAR LOS ARCHIVOS EXTRAÍDOS
# ------------------------------------------------------------

archivos_extraidos = []

for raiz, carpetas, archivos in os.walk(carpeta_extraccion):
    for archivo in archivos:
        ruta_completa = os.path.join(raiz, archivo)
        archivos_extraidos.append(ruta_completa)

print(f"Cantidad de archivos extraídos: {len(archivos_extraidos)}")

print("\nARCHIVOS EXTRAÍDOS")

for numero, ruta_completa in enumerate(
    sorted(archivos_extraidos),
    start=1
):
    ruta_relativa = os.path.relpath(
        ruta_completa,
        carpeta_extraccion
    )

    tamaño_mb = os.path.getsize(ruta_completa) / (1024 * 1024)

    print(
        f"{numero:03d}. "
        f"{ruta_relativa} | "
        f"{tamaño_mb:.2f} MB"
    )

# ------------------------------------------------------------
# VALIDACIÓN FINAL
# ------------------------------------------------------------

print("\n" + "=" * 90)

if len(archivos_extraidos) == 22:
    print("ESTADO: extracción correcta. Se encontraron los 22 archivos esperados.")
else:
    print(
        "ADVERTENCIA: la cantidad extraída no coincide con "
        "los 22 archivos esperados."
    )
```

```
# ============================================================
# PASO 6 - CLASIFICAR LOS ARCHIVOS POR GRUPO
# ============================================================

import os
import re
import pandas as pd

# ------------------------------------------------------------
# RUTA DE LOS ARCHIVOS EXTRAÍDOS
# ------------------------------------------------------------

carpeta_originales = "/content/01_ORIGINALES"

# ------------------------------------------------------------
# LISTAR TODOS LOS ARCHIVOS
# ------------------------------------------------------------

archivos = []

for raiz, carpetas, nombres_archivos in os.walk(carpeta_originales):
    for nombre_archivo in nombres_archivos:
        ruta_completa = os.path.join(raiz, nombre_archivo)

        archivos.append({
            "Archivo": nombre_archivo,
            "RutaCompleta": ruta_completa,
            "Extension": os.path.splitext(nombre_archivo)[1].lower()
        })

df_inventario = pd.DataFrame(archivos)

# ------------------------------------------------------------
# FUNCIÓN PARA CLASIFICAR CADA ARCHIVO
# ------------------------------------------------------------

def clasificar_archivo(nombre_archivo):

    nombre = nombre_archivo.lower()

    if "relacion cdp" in nombre or "relación cdp" in nombre:
        return "RELACION_CDP"

    elif "procesos por estado" in nombre:
        return "PROCESOS_ESTADO"

    elif re.search(r"\bpaa\s*20\d{2}", nombre):
        return "PAA"

    elif "ajustes y reservas" in nombre:
        return "AJUSTES_RESERVAS"

    elif "reporte de obras" in nombre:
        return "REPORTE_OBRAS"

    elif "reporte de pagos" in nombre:
        return "REPORTE_PAGOS"

    else:
        return "SIN_CLASIFICAR"

# ------------------------------------------------------------
# APLICAR CLASIFICACIÓN
# ------------------------------------------------------------

df_inventario["Grupo"] = df_inventario["Archivo"].apply(
    clasificar_archivo
)

# ------------------------------------------------------------
# EXTRAER AÑO DEL NOMBRE DEL ARCHIVO
# ------------------------------------------------------------

df_inventario["Anio"] = (
    df_inventario["Archivo"]
    .str.extract(r"(20\d{2})", expand=False)
)

# ------------------------------------------------------------
# ORDENAR RESULTADO
# ------------------------------------------------------------

df_inventario = df_inventario.sort_values(
    by=["Grupo", "Anio", "Archivo"]
).reset_index(drop=True)

# ------------------------------------------------------------
# MOSTRAR INVENTARIO
# ------------------------------------------------------------

print("=" * 100)
print("CLASIFICACIÓN DE ARCHIVOS")
print("=" * 100)

print(
    df_inventario[
        ["Grupo", "Anio", "Extension", "Archivo"]
    ].to_string(index=False)
)

# ------------------------------------------------------------
# RESUMEN POR GRUPO
# ------------------------------------------------------------

print("\n" + "=" * 100)
print("RESUMEN POR GRUPO")
print("=" * 100)

resumen = (
    df_inventario
    .groupby("Grupo")
    .size()
    .reset_index(name="CantidadArchivos")
)

print(resumen.to_string(index=False))

# ------------------------------------------------------------
# VALIDACIÓN FINAL
# ------------------------------------------------------------

sin_clasificar = df_inventario[
    df_inventario["Grupo"] == "SIN_CLASIFICAR"
]

print("\n" + "=" * 100)

if len(df_inventario) != 22:
    print(
        f"ADVERTENCIA: se encontraron {len(df_inventario)} archivos, "
        "pero se esperaban 22."
    )

elif not sin_clasificar.empty:
    print("ADVERTENCIA: existen archivos sin clasificar:")
    print(sin_clasificar["Archivo"].to_string(index=False))

else:
    print(
        "ESTADO: los 22 archivos fueron clasificados correctamente "
        "en sus grupos."
    )
```

```
# ============================================================
# PASO 7 - INSPECCIONAR LAS HOJAS DEL GRUPO RELACION_CDP
# ============================================================

import os
from openpyxl import load_workbook

# ------------------------------------------------------------
# SELECCIONAR ÚNICAMENTE LOS ARCHIVOS DEL GRUPO RELACION_CDP
# ------------------------------------------------------------

df_cdp = (
    df_inventario[
        df_inventario["Grupo"] == "RELACION_CDP"
    ]
    .sort_values("Anio")
    .reset_index(drop=True)
)

print("=" * 100)
print("INSPECCIÓN DE HOJAS - GRUPO RELACION_CDP")
print("=" * 100)

resultados_hojas = []
errores_hojas = []

# ------------------------------------------------------------
# ABRIR CADA ARCHIVO SIN CARGAR TODAS LAS CELDAS EN MEMORIA
# ------------------------------------------------------------

for _, fila in df_cdp.iterrows():

    archivo = fila["Archivo"]
    anio = fila["Anio"]
    ruta = fila["RutaCompleta"]

    print(f"\nArchivo: {archivo}")
    print(f"Año: {anio}")

    try:
        libro = load_workbook(
            filename=ruta,
            read_only=True,
            data_only=True
        )

        hojas = libro.worksheets

        print(f"Cantidad de hojas: {len(hojas)}")

        for numero, hoja in enumerate(hojas, start=1):

            estado = hoja.sheet_state

            print(
                f"  {numero}. "
                f"Hoja: {hoja.title} | "
                f"Estado: {estado}"
            )

            resultados_hojas.append({
                "Grupo": "RELACION_CDP",
                "Anio": anio,
                "Archivo": archivo,
                "NumeroHoja": numero,
                "NombreHoja": hoja.title,
                "EstadoHoja": estado
            })

        libro.close()

    except Exception as error:

        print(f"ERROR AL ABRIR EL ARCHIVO: {error}")

        errores_hojas.append({
            "Anio": anio,
            "Archivo": archivo,
            "Error": str(error)
        })

# ------------------------------------------------------------
# CREAR TABLA DE RESULTADOS
# ------------------------------------------------------------

df_hojas_cdp = pd.DataFrame(resultados_hojas)

print("\n" + "=" * 100)
print("RESUMEN DE HOJAS")
print("=" * 100)

if not df_hojas_cdp.empty:

    print(
        df_hojas_cdp[
            [
                "Anio",
                "NumeroHoja",
                "NombreHoja",
                "EstadoHoja",
                "Archivo"
            ]
        ].to_string(index=False)
    )

# ------------------------------------------------------------
# VALIDACIÓN FINAL
# ------------------------------------------------------------

print("\n" + "=" * 100)

if errores_hojas:

    print("ESTADO: algunos archivos presentaron errores.")

    for error in errores_hojas:
        print(
            f"- {error['Archivo']}: "
            f"{error['Error']}"
        )

elif len(df_cdp) != 4:

    print(
        f"ADVERTENCIA: se esperaban 4 archivos RELACION_CDP, "
        f"pero se encontraron {len(df_cdp)}."
    )

else:

    print(
        "ESTADO: los 4 archivos RELACION_CDP se abrieron "
        "correctamente y sus hojas fueron identificadas."
    )
```

```
# ============================================================
# PASO 9 - COMPARAR ENCABEZADOS DEL GRUPO RELACION_CDP
# ============================================================

import pandas as pd
import re
import unicodedata

# ------------------------------------------------------------
# FUNCIÓN DE NORMALIZACIÓN SOLO PARA COMPARAR
# NO MODIFICA TODAVÍA LOS NOMBRES ORIGINALES
# ------------------------------------------------------------

def normalizar_nombre_columna(nombre):

    nombre = str(nombre).strip()

    # Reemplazar saltos de línea por espacios
    nombre = nombre.replace("\n", " ")

    # Quitar tildes para facilitar la comparación
    nombre = unicodedata.normalize("NFKD", nombre)
    nombre = "".join(
        caracter
        for caracter in nombre
        if not unicodedata.combining(caracter)
    )

    # Convertir a minúsculas
    nombre = nombre.lower()

    # Reemplazar caracteres especiales por espacio
    nombre = re.sub(r"[^a-z0-9]+", " ", nombre)

    # Eliminar espacios repetidos
    nombre = re.sub(r"\s+", " ", nombre).strip()

    return nombre


print("=" * 120)
print("COMPARACIÓN DE COLUMNAS - RELACIÓN CDP")
print("=" * 120)

inventario_columnas = []
columnas_por_anio = {}
errores_columnas = []

# ------------------------------------------------------------
# LEER SOLO LOS ENCABEZADOS DE CADA ARCHIVO
# ------------------------------------------------------------

for _, fila in df_cdp.iterrows():

    anio = str(fila["Anio"])
    archivo = fila["Archivo"]
    ruta = fila["RutaCompleta"]
    hoja = hojas_principales_cdp[anio]

    print("\n" + "=" * 120)
    print(f"AÑO: {anio}")
    print(f"ARCHIVO: {archivo}")
    print(f"HOJA: {hoja}")
    print("=" * 120)

    try:

        encabezados = pd.read_excel(
            ruta,
            sheet_name=hoja,
            header=0,
            nrows=0,
            engine="openpyxl"
        )

        columnas_originales = list(encabezados.columns)

        columnas_por_anio[anio] = {
            normalizar_nombre_columna(columna)
            for columna in columnas_originales
        }

        print(f"Cantidad de columnas: {len(columnas_originales)}\n")

        for posicion, columna in enumerate(
            columnas_originales,
            start=1
        ):

            columna_normalizada = normalizar_nombre_columna(
                columna
            )

            print(
                f"{posicion:02d}. "
                f"Original: {columna} | "
                f"Normalizada: {columna_normalizada}"
            )

            inventario_columnas.append({
                "Anio": anio,
                "Archivo": archivo,
                "Posicion": posicion,
                "ColumnaOriginal": str(columna),
                "ColumnaNormalizada": columna_normalizada
            })

    except Exception as error:

        print(f"ERROR: {error}")

        errores_columnas.append({
            "Anio": anio,
            "Archivo": archivo,
            "Error": str(error)
        })

# ------------------------------------------------------------
# CREAR DATAFRAME DE INVENTARIO
# ------------------------------------------------------------

df_columnas_cdp = pd.DataFrame(inventario_columnas)

# ------------------------------------------------------------
# IDENTIFICAR COLUMNAS COMUNES A LOS CUATRO AÑOS
# ------------------------------------------------------------

print("\n" + "=" * 120)
print("COLUMNAS NORMALIZADAS COMUNES A LOS CUATRO AÑOS")
print("=" * 120)

if len(columnas_por_anio) == 4:

    columnas_comunes = set.intersection(
        *columnas_por_anio.values()
    )

    if columnas_comunes:

        for columna in sorted(columnas_comunes):
            print(f"- {columna}")

    else:
        print(
            "No existen columnas con exactamente el mismo "
            "nombre normalizado en los cuatro años."
        )

# ------------------------------------------------------------
# MOSTRAR COLUMNAS EXCLUSIVAS O DIFERENTES POR AÑO
# ------------------------------------------------------------

print("\n" + "=" * 120)
print("COLUMNAS NO COMUNES POR AÑO")
print("=" * 120)

if len(columnas_por_anio) == 4:

    for anio in sorted(columnas_por_anio):

        otros_anios = [
            columnas
            for otro_anio, columnas in columnas_por_anio.items()
            if otro_anio != anio
        ]

        union_otros = set.union(*otros_anios)

        exclusivas = (
            columnas_por_anio[anio]
            - union_otros
        )

        print(f"\nAño {anio}:")

        if exclusivas:

            for columna in sorted(exclusivas):
                print(f"- {columna}")

        else:
            print("- No tiene columnas exclusivas.")

# ------------------------------------------------------------
# MATRIZ DE PRESENCIA POR AÑO
# ------------------------------------------------------------

print("\n" + "=" * 120)
print("MATRIZ DE PRESENCIA DE COLUMNAS NORMALIZADAS")
print("=" * 120)

if columnas_por_anio:

    todas_las_columnas = sorted(
        set.union(*columnas_por_anio.values())
    )

    matriz_presencia = pd.DataFrame({
        "ColumnaNormalizada": todas_las_columnas
    })

    for anio in sorted(columnas_por_anio):

        matriz_presencia[anio] = (
            matriz_presencia["ColumnaNormalizada"]
            .isin(columnas_por_anio[anio])
            .map({
                True: "Sí",
                False: "No"
            })
        )

    print(
        matriz_presencia.to_string(
            index=False
        )
    )

# ------------------------------------------------------------
# VALIDACIÓN FINAL
# ------------------------------------------------------------

print("\n" + "=" * 120)

if errores_columnas:

    print(
        "ESTADO: se presentaron errores al leer "
        "los encabezados."
    )

    for error in errores_columnas:
        print(
            f"- Año {error['Anio']} | "
            f"{error['Archivo']} | "
            f"{error['Error']}"
        )

elif len(columnas_por_anio) != 4:

    print(
        "ADVERTENCIA: no se obtuvieron encabezados "
        "de los cuatro años."
    )

else:

    print(
        "ESTADO: los encabezados de los cuatro años "
        "fueron comparados correctamente."
    )
```

```
# ============================================================
# PASO 12 - IDENTIFICAR LAS FILAS REALES DE NEGOCIO
# GRUPO RELACION_CDP
# ============================================================

import pandas as pd
import numpy as np
import re
import unicodedata


# ------------------------------------------------------------
# FUNCIÓN PARA NORMALIZAR ENCABEZADOS SOLO PARA BÚSQUEDA
# ------------------------------------------------------------

def normalizar_encabezado(nombre):

    nombre = str(nombre).strip()
    nombre = nombre.replace("\n", " ")

    nombre = unicodedata.normalize("NFKD", nombre)

    nombre = "".join(
        caracter
        for caracter in nombre
        if not unicodedata.combining(caracter)
    )

    nombre = nombre.lower()
    nombre = re.sub(r"[^a-z0-9]+", " ", nombre)
    nombre = re.sub(r"\s+", " ", nombre).strip()

    return nombre


# ------------------------------------------------------------
# CAMPOS CLAVE POR AÑO
# ------------------------------------------------------------

campos_clave_cdp = {

    "2023": {
        "Fecha": "FECHA",
        "Proyecto": "PROYECTO PEP",
        "NumeroCDP": "No CDP"
    },

    "2024": {
        "Fecha": "FECHA CDP",
        "Proyecto": "PROYECTO PEP",
        "NumeroCDP": "No CDP"
    },

    "2025": {
        "Fecha": "Fecha solicitud CDP",
        "Proyecto": "Proyecto",
        "NumeroCDP": "N° CDP"
    },

    "2026": {
        "Fecha": "Fecha solicitud CDP",
        "Proyecto": "Proyecto",
        "NumeroCDP": "N° CDP"
    }
}


# ------------------------------------------------------------
# FUNCIÓN PARA LOCALIZAR UNA COLUMNA SIN DEPENDER DE ESPACIOS
# ------------------------------------------------------------

def encontrar_columna(df, nombre_buscado):

    nombre_normalizado = normalizar_encabezado(nombre_buscado)

    coincidencias = [
        columna
        for columna in df.columns
        if normalizar_encabezado(columna) == nombre_normalizado
    ]

    if not coincidencias:
        return None

    return coincidencias[0]


# ------------------------------------------------------------
# CONTENEDORES
# ------------------------------------------------------------

resumen_filas_reales = []
filas_parciales_detectadas = []
datos_reales_por_anio = {}
errores_filas_reales = []


print("=" * 130)
print("IDENTIFICACIÓN DE FILAS REALES DE NEGOCIO - RELACIÓN CDP")
print("=" * 130)


# ------------------------------------------------------------
# ANALIZAR CADA AÑO
# ------------------------------------------------------------

for _, fila_inventario in df_cdp.iterrows():

    anio = str(fila_inventario["Anio"])
    archivo = fila_inventario["Archivo"]

    print("\n" + "=" * 130)
    print(f"AÑO: {anio}")
    print(f"ARCHIVO: {archivo}")
    print("=" * 130)

    try:

        # Utilizamos el DataFrame ya leído en el Paso 11
        df_anio = datos_por_anio[anio].copy()

        # Mantener el número real de fila del Excel
        # Encabezado = fila 1, primer dato = fila 2
        df_anio["_FilaExcel"] = df_anio.index + 2

        columna_fecha = encontrar_columna(
            df_anio,
            campos_clave_cdp[anio]["Fecha"]
        )

        columna_proyecto = encontrar_columna(
            df_anio,
            campos_clave_cdp[anio]["Proyecto"]
        )

        columna_cdp = encontrar_columna(
            df_anio,
            campos_clave_cdp[anio]["NumeroCDP"]
        )

        if not all([
            columna_fecha,
            columna_proyecto,
            columna_cdp
        ]):
            raise ValueError(
                "No fue posible localizar todos los campos clave."
            )

        # ----------------------------------------------------
        # LIMPIAR CADENAS VACÍAS EN LOS CAMPOS CLAVE
        # ----------------------------------------------------

        for columna in [
            columna_fecha,
            columna_proyecto,
            columna_cdp
        ]:

            df_anio[columna] = df_anio[columna].replace(
                r"^\s*$",
                np.nan,
                regex=True
            )

        # ----------------------------------------------------
        # INDICADORES DE PRESENCIA
        # ----------------------------------------------------

        tiene_fecha = df_anio[columna_fecha].notna()
        tiene_proyecto = df_anio[columna_proyecto].notna()
        tiene_cdp = df_anio[columna_cdp].notna()

        cantidad_claves = (
            tiene_fecha.astype(int)
            + tiene_proyecto.astype(int)
            + tiene_cdp.astype(int)
        )

        # Registro candidato:
        # tiene al menos uno de los tres campos clave
        mascara_registro_candidato = cantidad_claves >= 1

        df_candidatos = (
            df_anio.loc[mascara_registro_candidato]
            .copy()
        )

        # Registro completo:
        # tiene los tres campos clave
        mascara_registro_completo = cantidad_claves == 3

        # Registro parcial:
        # tiene uno o dos campos clave
        mascara_registro_parcial = (
            (cantidad_claves >= 1)
            &
            (cantidad_claves < 3)
        )

        df_parciales = (
            df_anio.loc[
                mascara_registro_parcial,
                [
                    "_FilaExcel",
                    columna_fecha,
                    columna_proyecto,
                    columna_cdp
                ]
            ]
            .copy()
        )

        # Guardar candidatos para pasos posteriores,
        # pero todavía no se consolidan
        datos_reales_por_anio[anio] = df_candidatos.copy()

        cantidad_original = len(df_anio)
        cantidad_candidatos = len(df_candidatos)
        cantidad_completos = int(
            mascara_registro_completo.sum()
        )
        cantidad_parciales = int(
            mascara_registro_parcial.sum()
        )
        cantidad_residuales = (
            cantidad_original - cantidad_candidatos
        )

        primera_fila_real = (
            int(df_candidatos["_FilaExcel"].min())
            if not df_candidatos.empty
            else None
        )

        ultima_fila_real = (
            int(df_candidatos["_FilaExcel"].max())
            if not df_candidatos.empty
            else None
        )

        print(f"Filas leídas originalmente: {cantidad_original}")
        print(f"Registros candidatos: {cantidad_candidatos}")
        print(f"Registros con las 3 claves: {cantidad_completos}")
        print(f"Registros parciales: {cantidad_parciales}")
        print(f"Filas residuales descartables: {cantidad_residuales}")
        print(f"Primera fila real del Excel: {primera_fila_real}")
        print(f"Última fila real del Excel: {ultima_fila_real}")

        # ----------------------------------------------------
        # DISTRIBUCIÓN POR COMBINACIÓN DE CAMPOS CLAVE
        # ----------------------------------------------------

        combinaciones = pd.DataFrame({
            "TieneFecha": tiene_fecha,
            "TieneProyecto": tiene_proyecto,
            "TieneNumeroCDP": tiene_cdp
        })

        resumen_combinaciones = (
            combinaciones
            .value_counts()
            .reset_index(name="CantidadFilas")
            .sort_values(
                "CantidadFilas",
                ascending=False
            )
        )

        print("\nCOMBINACIONES DE CAMPOS CLAVE")

        print(
            resumen_combinaciones.to_string(
                index=False
            )
        )

        # ----------------------------------------------------
        # MOSTRAR FILAS PARCIALES
        # ----------------------------------------------------

        print("\nREGISTROS PARCIALES")

        if df_parciales.empty:

            print(
                "No se encontraron registros parciales. "
                "Todos los candidatos tienen las tres claves."
            )

        else:

            print(
                df_parciales.head(30).to_string(
                    index=False
                )
            )

            for _, fila_parcial in df_parciales.iterrows():

                filas_parciales_detectadas.append({
                    "Anio": anio,
                    "Archivo": archivo,
                    "FilaExcel": fila_parcial["_FilaExcel"],
                    "Fecha": fila_parcial[columna_fecha],
                    "Proyecto": fila_parcial[columna_proyecto],
                    "NumeroCDP": fila_parcial[columna_cdp]
                })

        # ----------------------------------------------------
        # MOSTRAR LAS ÚLTIMAS FILAS CANDIDATAS
        # ----------------------------------------------------

        print("\nÚLTIMAS 10 FILAS CANDIDATAS")

        columnas_muestra = [
            "_FilaExcel",
            columna_fecha,
            columna_proyecto,
            columna_cdp
        ]

        print(
            df_candidatos[
                columnas_muestra
            ]
            .tail(10)
            .to_string(index=False)
        )

        resumen_filas_reales.append({
            "Anio": anio,
            "Archivo": archivo,
            "FilasOriginales": cantidad_original,
            "RegistrosCandidatos": cantidad_candidatos,
            "RegistrosCompletos": cantidad_completos,
            "RegistrosParciales": cantidad_parciales,
            "FilasResiduales": cantidad_residuales,
            "PrimeraFilaExcel": primera_fila_real,
            "UltimaFilaExcel": ultima_fila_real
        })

    except Exception as error:

        print(f"ERROR: {error}")

        errores_filas_reales.append({
            "Anio": anio,
            "Archivo": archivo,
            "Error": str(error)
        })


# ------------------------------------------------------------
# RESUMEN GENERAL
# ------------------------------------------------------------

df_resumen_filas_reales = pd.DataFrame(
    resumen_filas_reales
)

df_filas_parciales_cdp = pd.DataFrame(
    filas_parciales_detectadas
)

print("\n" + "=" * 130)
print("RESUMEN GENERAL DE FILAS REALES")
print("=" * 130)

if not df_resumen_filas_reales.empty:

    print(
        df_resumen_filas_reales.to_string(
            index=False
        )
    )


# ------------------------------------------------------------
# VALIDACIÓN FINAL
# ------------------------------------------------------------

print("\n" + "=" * 130)

if errores_filas_reales:

    print(
        "ESTADO: se presentaron errores al identificar "
        "las filas reales."
    )

    for error in errores_filas_reales:

        print(
            f"- Año {error['Anio']} | "
            f"{error['Archivo']} | "
            f"{error['Error']}"
        )

elif len(datos_reales_por_anio) != 4:

    print(
        "ADVERTENCIA: no se analizaron correctamente "
        "los cuatro años."
    )

else:

    print(
        "ESTADO: se identificaron las filas candidatas "
        "de los cuatro años. Todavía no se eliminaron, "
        "renombraron ni unieron datos."
    )
```
